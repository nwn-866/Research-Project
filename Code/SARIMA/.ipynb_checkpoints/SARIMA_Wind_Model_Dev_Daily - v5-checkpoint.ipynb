```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test on the EWMA-smoothed series
result = adfuller(df_model_reduce_noice['Total_TP_EWMA'])

print("ADF Statistic:", result[0])
print("p-value:", result[1])
print("Critical Values:")
for key, value in result[4].items():
    print(f"\t{key}: {value:.3f}")
```

```python
# -----------------------------
# 2. Define Exogenous Variables
# -----------------------------
# For example, we will use PS, T2M, and WS50M as exogenous predictors.
exog_columns = ['WS50M']
# (You can add more predictors if desired, e.g. wind quadrant dummies, if they vary meaningfully)

# -----------------------------
# 3. Train-Test Split
# -----------------------------
# Choose a cutoff date for splitting the data. For instance, training data up to 2024-01-31,
# and testing data from 2024-02-01 onward.
train_end = '2024-01-31'
train = df_model_reduce_noice.loc[:train_end]
test = df_model_reduce_noice.loc['2024-02-01':]

print("Train set:", train.index.min(), "to", train.index.max())
print("Test set:", test.index.min(), "to", test.index.max())

# Define the target variable as the EWMA-smoothed series
target_train = train['Total_TP_EWMA']
target_test = test['Total_TP_EWMA']

# Define exogenous variables (using the original values)
exog_train = train[exog_columns]
exog_test = test[exog_columns]

# -----------------------------
# 4. Fit a Daily SARIMAX Model
# -----------------------------
# For this example, we use an ARIMA(1,1,1) model without seasonal terms.
model = SARIMAX(target_train,
                exog=exog_train,
                order=(1, 1, 1),
                seasonal_order=(0, 0, 0, 0),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)
print(results.summary())

# -----------------------------
# 5. Forecasting and Performance Metrics
# -----------------------------
n_steps = len(test)
forecast = results.get_forecast(steps=n_steps, exog=exog_test).predicted_mean

# Compute performance metrics
rmse = np.sqrt(mean_squared_error(target_test, forecast))
mae = mean_absolute_error(target_test, forecast)
epsilon = 1e-10  # to avoid division by zero in MAPE calculation
mape = np.mean(np.abs((target_test - forecast) / (target_test + epsilon))) * 100

print("\nForecast Performance Metrics:")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"Adjusted MAPE: {mape:.2f}%")
```

```python
# Define target and exogenous variables:
target = df_model_reduce_noice['Total_TP_EWMA']
exog = df_model_reduce_noice[['WS50M']]  # You can add more predictors if desired

# Run auto_arima to identify the best orders
auto_model = pm.auto_arima(
    target,
    exogenous=exog,
    seasonal=True,    # We're testing for seasonal patterns
    m=7,              # For daily data, m=7 for weekly seasonality; adjust if needed.
    start_p=0, max_p=3,
    start_q=0, max_q=3,
    d=0,              # d=0 because the EWMA-smoothed series is stationary (as per the ADF test)
    start_P=0, max_P=1,
    start_Q=0, max_Q=1,
    D=0,              # We assume no seasonal differencing is needed given the stationarity
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True
)

print("auto_arima selected order:", auto_model.order)
print("auto_arima selected seasonal_order:", auto_model.seasonal_order)

# Optionally, visualize the model diagnostics from auto_arima:
auto_model.plot_diagnostics(figsize=(12, 8))
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error

# --- Train-Test Split ---
# For example, training from 2015-03-01 to 2024-01-31, test from 2024-02-01 to 2025-01-31.
train_end = '2024-01-31'
train = df_model_selected.loc[:train_end]
test = df_model_selected.loc['2024-02-01':]

print("Train set:", train.index.min(), "to", train.index.max())
print("Test set:", test.index.min(), "to", test.index.max())

# Define target variables (in original scale)
target_train = train['Total_TP']
target_test = test['Total_TP']

# -----------------------------------------------
# 1. Baseline Model (Without Exogenous Variables)
# -----------------------------------------------
model_baseline = SARIMAX(target_train,
                         order=(3, 1, 2),
                         seasonal_order=(0, 0, 0, 7),
                         enforce_stationarity=False,
                         enforce_invertibility=False,
                         missing='drop')
results_baseline = model_baseline.fit(disp=False)
print("Baseline Model Summary (Without Log Transform):")
print(results_baseline.summary())

# Forecast for the test period
n_steps = len(test)
forecast_baseline = results_baseline.get_forecast(steps=n_steps).predicted_mean

# Compute performance metrics for the baseline model
rmse_baseline = np.sqrt(mean_squared_error(target_test, forecast_baseline))
mae_baseline = mean_absolute_error(target_test, forecast_baseline)
epsilon = 1e-10  # small number to avoid division by zero
mape_baseline = np.mean(np.abs((target_test - forecast_baseline) / (target_test + epsilon))) * 100

print("\nBaseline Forecast Performance Metrics (Without Log Transform):")
print(f"RMSE: {rmse_baseline:.2f}")
print(f"MAE: {mae_baseline:.2f}")
print(f"Adjusted MAPE: {mape_baseline:.2f}%")
```

```python
# Define a function to perform the ADF test for stationarity
def adf_test(series, title=''):
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '# Lags Used', 'Number of Observations Used']
    for value, label in zip(result, labels):
        print(f'{label} : {value}')
    if result[1] <= 0.05:
        print("=> The series is likely stationary (reject H0).")
    else:
        print("=> The series is likely non-stationary (fail to reject H0).")
```
