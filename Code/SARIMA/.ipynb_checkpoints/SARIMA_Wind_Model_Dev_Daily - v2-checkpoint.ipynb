```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error

# --- Train-Test Split ---
# For example, training from 2015-03-01 to 2024-01-31, test from 2024-02-01 to 2025-01-31.
train_end = '2024-01-31'
train = df_model_selected.loc[:train_end]
test = df_model_selected.loc['2024-02-01':]

print("Train set:", train.index.min(), "to", train.index.max())
print("Test set:", test.index.min(), "to", test.index.max())

# Define target variables (in original scale)
target_train = train['Total_TP']
target_test = test['Total_TP']

# -----------------------------------------------
# 1. Baseline Model (Without Exogenous Variables)
# -----------------------------------------------
model_baseline = SARIMAX(target_train,
                         order=(3, 1, 2),
                         seasonal_order=(0, 0, 0, 7),
                         enforce_stationarity=False,
                         enforce_invertibility=False,
                         missing='drop')
results_baseline = model_baseline.fit(disp=False)
print("Baseline Model Summary (Without Log Transform):")
print(results_baseline.summary())

# Forecast for the test period
n_steps = len(test)
forecast_baseline = results_baseline.get_forecast(steps=n_steps).predicted_mean

# Compute performance metrics for the baseline model
rmse_baseline = np.sqrt(mean_squared_error(target_test, forecast_baseline))
mae_baseline = mean_absolute_error(target_test, forecast_baseline)
epsilon = 1e-10  # small number to avoid division by zero
mape_baseline = np.mean(np.abs((target_test - forecast_baseline) / (target_test + epsilon))) * 100

print("\nBaseline Forecast Performance Metrics (Without Log Transform):")
print(f"RMSE: {rmse_baseline:.2f}")
print(f"MAE: {mae_baseline:.2f}")
print(f"Adjusted MAPE: {mape_baseline:.2f}%")
```

```python
# Resample daily data to weekly frequency using mean aggregation
df_weekly = df_model_selected.resample('W').mean()

print("Weekly Data Range:", df_weekly.index.min(), "to", df_weekly.index.max())

# --- Train-Test Split on Weekly Data ---
# For example, let's use training data until 2024-01-31, and test data from 2024-02-01 onward.
train_end = '2024-01-31'
train_weekly = df_weekly.loc[:train_end]
test_weekly = df_weekly.loc['2024-02-01':]

print("Weekly Train set:", train_weekly.index.min(), "to", train_weekly.index.max())
print("Weekly Test set:", test_weekly.index.min(), "to", test_weekly.index.max())

# Define the target variable for the weekly data
target_train_weekly = train_weekly['Total_TP']
target_test_weekly = test_weekly['Total_TP']

# --- Fit Baseline SARIMAX Model on Weekly Data ---
# We'll use ARIMA(3,1,2) without a seasonal component (seasonal_order=(0,0,0,0))
model_weekly = SARIMAX(target_train_weekly, 
                       order=(1, 1, 2), 
                       seasonal_order=(0, 0, 0, 0),
                       enforce_stationarity=False, 
                       enforce_invertibility=False)
results_weekly = model_weekly.fit(disp=False)
print(results_weekly.summary())

# --- Forecasting ---
n_steps_weekly = len(test_weekly)
forecast_weekly = results_weekly.get_forecast(steps=n_steps_weekly).predicted_mean

# --- Compute Performance Metrics ---
rmse_weekly = np.sqrt(mean_squared_error(target_test_weekly, forecast_weekly))
mae_weekly = mean_absolute_error(target_test_weekly, forecast_weekly)
epsilon = 1e-10  # small constant to avoid division by zero
mape_weekly = np.mean(np.abs((target_test_weekly - forecast_weekly) / (target_test_weekly + epsilon))) * 100

print("Weekly Baseline Model Forecast Performance Metrics:")
print(f"RMSE: {rmse_weekly:.2f}")
print(f"MAE: {mae_weekly:.2f}")
print(f"Adjusted MAPE: {mape_weekly:.2f}%")
```

```python
# Define target variable and exogenous variables for weekly data
# target_train_weekly = train_weekly['Total_TP']
# target_test_weekly = test_weekly['Total_TP']
exog_train_weekly = train_weekly[['u_scaled', 'v_scaled']]
exog_test_weekly = test_weekly[['u_scaled', 'v_scaled']]

# --- Fit SARIMAX Model on Weekly Data with Exogenous Variables ---
# We'll use the same non-seasonal order as before, ARIMA(3,1,2), and no seasonal terms.
model_weekly_exog = SARIMAX(target_train_weekly, exog=exog_train_weekly,
                            order=(3, 1, 2), seasonal_order=(0, 0, 0, 0),
                            enforce_stationarity=False, enforce_invertibility=False)
results_weekly_exog = model_weekly_exog.fit(disp=False)
print(results_weekly_exog.summary())

# --- Forecasting ---
n_steps_weekly = len(test_weekly)
forecast_weekly_exog = results_weekly_exog.get_forecast(steps=n_steps_weekly, exog=exog_test_weekly).predicted_mean

# --- Compute Performance Metrics ---
rmse_weekly_exog = np.sqrt(mean_squared_error(target_test_weekly, forecast_weekly_exog))
mae_weekly_exog = mean_absolute_error(target_test_weekly, forecast_weekly_exog)
epsilon = 1e-10  # small constant to avoid division by zero in MAPE
mape_weekly_exog = np.mean(np.abs((target_test_weekly - forecast_weekly_exog) / (target_test_weekly + epsilon))) * 100

print("\nWeekly Exogenous Model Forecast Performance Metrics:")
print(f"RMSE: {rmse_weekly_exog:.2f}")
print(f"MAE: {mae_weekly_exog:.2f}")
print(f"Adjusted MAPE: {mape_weekly_exog:.2f}%")
```

```python
# Define exogenous variables for weekly data: now include u_scaled, v_scaled, and PS
exog_train_weekly = train_weekly[['u_scaled', 'v_scaled', 'PS','T2M']]
exog_test_weekly = test_weekly[['u_scaled', 'v_scaled', 'PS','T2M']]

# --- Fit SARIMAX Model on Weekly Data with Exogenous Variables (u_scaled, v_scaled, PS) ---
# We'll use ARIMA(3,1,2) with no seasonal component (seasonal_order=(0,0,0,0)) for the weekly data.
model_weekly_exog = SARIMAX(target_train_weekly,
                            exog=exog_train_weekly,
                            order=(1, 1, 2),
                            seasonal_order=(0, 0, 0, 0),
                            enforce_stationarity=False,
                            enforce_invertibility=False)
results_weekly_exog = model_weekly_exog.fit(disp=False)
print(results_weekly_exog.summary())

# --- Forecasting ---
n_steps_weekly = len(test_weekly)
forecast_weekly_exog = results_weekly_exog.get_forecast(steps=n_steps_weekly,
                                                         exog=exog_test_weekly).predicted_mean

# --- Compute Performance Metrics ---
rmse_weekly_exog = np.sqrt(mean_squared_error(target_test_weekly, forecast_weekly_exog))
mae_weekly_exog = mean_absolute_error(target_test_weekly, forecast_weekly_exog)
epsilon = 1e-10  # Small constant to avoid division by zero
mape_weekly_exog = np.mean(np.abs((target_test_weekly - forecast_weekly_exog) / (target_test_weekly + epsilon))) * 100

print("\nWeekly Exogenous Model Forecast Performance Metrics (with PS):")
print(f"RMSE: {rmse_weekly_exog:.2f}")
print(f"MAE: {mae_weekly_exog:.2f}")
print(f"Adjusted MAPE: {mape_weekly_exog:.2f}%")
```

```python
# Define a function to perform the ADF test for stationarity
def adf_test(series, title=''):
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '# Lags Used', 'Number of Observations Used']
    for value, label in zip(result, labels):
        print(f'{label} : {value}')
    if result[1] <= 0.05:
        print("=> The series is likely stationary (reject H0).")
    else:
        print("=> The series is likely non-stationary (fail to reject H0).")
```

```python
# Run the ADF test on Total_TP
adf_test(df_model_selected['Total_TP'], title='Total_TP')
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Feature Engineering ---
weekly_df['WS50M_lag1'] = weekly_df['Avg_WS50M'].shift(1)
weekly_df['WD50M_lag1'] = weekly_df['Avg_WD50M'].shift(1)
weekly_df['WS50M_rolling3'] = weekly_df['Avg_WS50M'].rolling(3).mean()
weekly_df.dropna(inplace=True)

# --- Step 2: Prepare exogenous variables ---
exog_vars = ['Avg_WS50M', 'Avg_WD50M', 'WS50M_lag1', 'WD50M_lag1', 'WS50M_rolling3']

# --- Step 3: Set index to datetime ---
weekly_df.index = pd.to_datetime(weekly_df.index)

# --- Step 4: Split into train and test ---
train = weekly_df[:'2023-12-31']
test = weekly_df['2024-01-01':]

y_train = train['Total_TP']
y_test = test['Total_TP']

exog_train_raw = train[exog_vars]
exog_test_raw = test[exog_vars]

# --- Step 5: Normalize exogenous variables ---
scaler = StandardScaler()
exog_train = pd.DataFrame(scaler.fit_transform(exog_train_raw),
                          columns=exog_vars, index=exog_train_raw.index)
exog_test = pd.DataFrame(scaler.transform(exog_test_raw),
                         columns=exog_vars, index=exog_test_raw.index)
```

```python
# Fit the SARIMAX model with best parameters
model = SARIMAX(y_train,
                exog=exog_train,
                order=(2, 1, 3),
                seasonal_order=(2, 0, 1, 7),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)
print(results.summary())
```

```python
# --- Step 8: Forecast ---
forecast = results.predict(start=y_test.index[0],
                           end=y_test.index[-1],
                           exog=exog_test)

# --- Step 9: Evaluate ---
mae = mean_absolute_error(y_test, forecast)
rmse = np.sqrt(mean_squared_error(y_test, forecast))

print(f"\n✅ MAE: {mae:,.2f}")
print(f"✅ RMSE: {rmse:,.2f}")

# --- Step 10: Plot Actual vs Forecast ---
plt.figure(figsize=(14, 6))
plt.plot(y_test.index, y_test, label='Actual', linewidth=2)
plt.plot(forecast.index, forecast, label='Forecast', linestyle='--', color='orange')
plt.title('Forecast vs Actual Wind Energy (2024)\nUsing Avg_WS50M, Avg_WD50M, Lags & Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Total_TP (Wind Energy)')
plt.legend()
plt.tight_layout()
plt.show()
```
