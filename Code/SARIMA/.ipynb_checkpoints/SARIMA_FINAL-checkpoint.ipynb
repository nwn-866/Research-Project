```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python

# 1: Load all datasets 
wind_data = pd.read_csv('wind_data.csv')
avg_ps = pd.read_csv('1Wind_Factor_Daily_AVG_PS.csv')
avg_t2m = pd.read_csv('2Wind_Factor_Daily_AVG_T2M.csv')
avg_wd50m = pd.read_csv('3Wind_Factor_Daily_AVG_WD50M.csv')
avg_ws50m = pd.read_csv('4Wind_Factor_Daily_AVG_WS50M_Formatted.csv')

# 2: Convert DATE columns to datetime (fixing format issues) 
wind_data['Trading_date'] = pd.to_datetime(wind_data['Trading_date'])
avg_ps['DATE'] = pd.to_datetime(avg_ps['DATE'])
avg_t2m['DATE'] = pd.to_datetime(avg_t2m['DATE'], dayfirst=True) 
avg_wd50m['DATE'] = pd.to_datetime(avg_wd50m['DATE'])
avg_ws50m['DATE'] = pd.to_datetime(avg_ws50m['DATE'])

# Rename Trading_date to DATE for merging
wind_data = wind_data.rename(columns={'Trading_date': 'DATE'})

wind_data
```

```python
# 3: Merge all datasets on DATE 
merged_df = wind_data[['DATE', 'Total_TP']] \
    .merge(avg_ps, on='DATE') \
    .merge(avg_t2m, on='DATE') \
    .merge(avg_wd50m, on='DATE') \
    .merge(avg_ws50m, on='DATE')

#  4: Set DATE as index 
merged_df.set_index('DATE', inplace=True)

# 5: Resample to weekly frequency (mean per week) 
# Note: resampling handles leap years automatically
weekly_df = merged_df.resample('W').mean()

#  Check result 
print("Weekly data shape:", weekly_df.shape)
print(weekly_df.head())
```

```python
weekly_df
```

```python
#   7  Plot Total_TP over time (Weekly) 
plt.figure(figsize=(12, 5))
plt.plot(weekly_df['Total_TP'], label='Weekly Wind Energy', marker='o', linestyle='-')
plt.title('Weekly Wind Renewable Energy Generation (2015–2024)')
plt.xlabel('Date')
plt.ylabel('Total_TP (Mean per Week)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#   8 Correlation matrix using weekly data 
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
corr = weekly_df.corr()  # Use weekly_df instead of merged_df
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix (Weekly Data)')
plt.tight_layout()
plt.show()
```

```python
#   9 Plot each climate factor (weekly) 
factors = ['Avg_PS', 'Avg_T2M', 'Avg_WD50M', 'Avg_WS50M']

for factor in factors:
    plt.figure(figsize=(12, 4))
    plt.plot(weekly_df[factor], label=factor, marker='o', linestyle='-')
    plt.title(f'{factor} Over Time (Weekly Mean)')
    plt.xlabel('Date')
    plt.ylabel(factor)
    plt.legend()
    plt.tight_layout()
    plt.show()

```

```python
from statsmodels.tsa.stattools import adfuller

# 10: Augmented Dickey-Fuller (ADF) Test on weekly data 
print("\nAugmented Dickey-Fuller Test on 'Total_TP' (Weekly):")
adf_result = adfuller(weekly_df['Total_TP'].dropna())

print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'   {key}: {value}')
```

ADF Statistic (-1.1871) is higher than the critical value (-2.8672 at 5%), Since fail to reject the null hypothesis, meaning the series likely has a trend or other non-stationary behavior.

And the p-value (0.6791) is more greater than 0.05. A high p-value (> 0.05) suggests that the null hypothesis cannot be rejected, confirming that the time series is non-stationary.

So the next step is "Differencing the Time Series"

```python
weekly_diff = weekly_df['Total_TP'].diff().dropna()
```

```python
from statsmodels.tsa.stattools import adfuller

print("\nADF Test on Differenced Weekly 'Total_TP':")
adf_result = adfuller(weekly_diff)

print(f"ADF Statistic: {adf_result[0]}")
print(f"p-value: {adf_result[1]}")
print("Critical Values:")
for key, value in adf_result[4].items():
    print(f"   {key}: {value}")
```

The ADF Statistic (-12.84) is much lower than the critical value (-2.8672)
Now we can proceed with time series modeling since the data is stationary



```python
# Make sure DATE is datetime and set as index (already handled during resampling)
# weekly_df should already have date time index after resampling

#  Split weekly data by date 
train = weekly_df[:'2023-12-31']
test = weekly_df['2024-01-01':]

#  Define exogenous variables 
exog_vars = ['Avg_PS', 'Avg_T2M', 'Avg_WD50M', 'Avg_WS50M']
exog_train = train[exog_vars]
exog_test = test[exog_vars]

#  Define target variable 
y_train = train['Total_TP']
y_test = test['Total_TP']

#  Print shapes 
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
from pmdarima import auto_arima


auto_model = auto_arima(y_train,
                        exogenous=exog_train,
                        seasonal=True,
                        m=7,  # weekly seasonality (cycle every 7 weeks)
                        trace=True,
                        stepwise=True)

print(auto_model.summary())
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit the SARIMAX model with best parameters
model = SARIMAX(y_train,
                exog=exog_train,
                order=(2, 1, 3),
                seasonal_order=(2, 0, 1, 7),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)
print(results.summary())
```

```python
forecast = results.predict(start=test.index[0],
                           end=test.index[-1],
                           exog=exog_test)
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(y_test.index, y_test, label='Actual', linewidth=2)
plt.plot(forecast.index, forecast, label='Forecast', linestyle='--')
plt.title('Weekly Forecast vs Actual Wind Energy (2024)')
plt.xlabel('Date')
plt.ylabel('Total_TP (Wind Energy)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, forecast)
rmse = np.sqrt(mean_squared_error(y_test, forecast))

print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

#Prepare exogenous variables
exog_vars = ['Avg_WS50M', 'Avg_WD50M']

# Add weekly index
weekly_df.index = pd.to_datetime(weekly_df.index)

# Split train/test 
train = weekly_df[:'2023-12-31']
test = weekly_df['2024-01-01':]

y_train = train['Total_TP']
y_test = test['Total_TP']

exog_train_raw = train[exog_vars]
exog_test_raw = test[exog_vars]

# Normalize exogenous variables
scaler = StandardScaler()

exog_train = pd.DataFrame(scaler.fit_transform(exog_train_raw),
                          columns=exog_vars, index=exog_train_raw.index)

exog_test = pd.DataFrame(scaler.transform(exog_test_raw),
                         columns=exog_vars, index=exog_test_raw.index)
```

```python
#  auto_arima to find best order
auto_model = auto_arima(y_train,
                        exogenous=exog_train,
                        seasonal=True,
                        m=7,  
                        trace=True,
                        stepwise=True,
                        suppress_warnings=True)
```

```python
# Fit SARIMAX model using best parameters
best_order = auto_model.order
best_seasonal_order = auto_model.seasonal_order

model = SARIMAX(y_train,
                exog=exog_train,
                order=best_order,
                seasonal_order=best_seasonal_order,
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)
print(results.summary())

# Forecast for test period (2024)
forecast = results.predict(start=y_test.index[0],
                           end=y_test.index[-1],
                           exog=exog_test)

# Evaluate performance
mae = mean_absolute_error(y_test, forecast)
rmse = np.sqrt(mean_squared_error(y_test, forecast))

print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")


# Plot predict vs actual 
plt.figure(figsize=(14, 6))
plt.plot(y_test.index, y_test, label='Actual', linewidth=2)
plt.plot(forecast.index, forecast, label='Forecast', linestyle='--')
plt.title('Forecast vs Actual Wind Energy (2024)\nUsing Avg_WS50M & Avg_WD50M')
plt.xlabel('Date')
plt.ylabel('Total_TP (Wind Energy)')
plt.legend()
plt.tight_layout()
plt.show()
```

# Here added Feature Engineering part

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Feature Engineering ---
weekly_df['WS50M_lag1'] = weekly_df['Avg_WS50M'].shift(1)
weekly_df['WD50M_lag1'] = weekly_df['Avg_WD50M'].shift(1)
weekly_df['WS50M_rolling3'] = weekly_df['Avg_WS50M'].rolling(3).mean()
weekly_df.dropna(inplace=True)

# --- Step 2: Prepare exogenous variables ---
exog_vars = ['Avg_WS50M', 'Avg_WD50M', 'WS50M_lag1', 'WD50M_lag1', 'WS50M_rolling3']

# --- Step 3: Set index to datetime ---
weekly_df.index = pd.to_datetime(weekly_df.index)

# --- Step 4: Split into train and test ---
train = weekly_df[:'2023-12-31']
test = weekly_df['2024-01-01':]

y_train = train['Total_TP']
y_test = test['Total_TP']

exog_train_raw = train[exog_vars]
exog_test_raw = test[exog_vars]

# --- Step 5: Normalize exogenous variables ---
scaler = StandardScaler()
exog_train = pd.DataFrame(scaler.fit_transform(exog_train_raw),
                          columns=exog_vars, index=exog_train_raw.index)
exog_test = pd.DataFrame(scaler.transform(exog_test_raw),
                         columns=exog_vars, index=exog_test_raw.index)
```

```python
# Fit the SARIMAX model with best parameters
model = SARIMAX(y_train,
                exog=exog_train,
                order=(2, 1, 3),
                seasonal_order=(2, 0, 1, 7),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False)
print(results.summary())
```

```python
# --- Step 8: Forecast ---
forecast = results.predict(start=y_test.index[0],
                           end=y_test.index[-1],
                           exog=exog_test)

# --- Step 9: Evaluate ---
mae = mean_absolute_error(y_test, forecast)
rmse = np.sqrt(mean_squared_error(y_test, forecast))

print(f"\n✅ MAE: {mae:,.2f}")
print(f"✅ RMSE: {rmse:,.2f}")

# --- Step 10: Plot Actual vs Forecast ---
plt.figure(figsize=(14, 6))
plt.plot(y_test.index, y_test, label='Actual', linewidth=2)
plt.plot(forecast.index, forecast, label='Forecast', linestyle='--', color='orange')
plt.title('Forecast vs Actual Wind Energy (2024)\nUsing Avg_WS50M, Avg_WD50M, Lags & Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Total_TP (Wind Energy)')
plt.legend()
plt.tight_layout()
plt.show()
```
