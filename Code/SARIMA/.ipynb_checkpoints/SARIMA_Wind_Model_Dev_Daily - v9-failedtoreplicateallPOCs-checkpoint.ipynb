# Train the model 

```python
# Create a log-transformed target. Adding 1 to avoid log(0) if needed.
df_WWD1102_weekly['Total_TP_log'] = np.log(df_WWD1102_weekly['Total_TP'] + 1)
```

```python
# Split the weekly data into 80% training and 20% testing sets
n = len(df_WWD1102_weekly)
train_size = int(n * 0.8)
train_weekly = df_WWD1102_weekly.iloc[:train_size]
test_weekly = df_WWD1102_weekly.iloc[train_size:]
```

```python
print("Weekly Train set:", train_weekly.index.min(), "to", train_weekly.index.max())
print("Weekly Test set:", test_weekly.index.min(), "to", test_weekly.index.max())
```

```python
# Define the target variable (original Total_TP) and exogenous variables
target_train_log = train_weekly['Total_TP_log']
target_test_log = test_weekly['Total_TP_log']

target_train = train_weekly['Total_TP']
target_test = test_weekly['Total_TP']

exog_cols = ['WS50M','PS']
exog_train = train_weekly[exog_cols]
exog_test = test_weekly[exog_cols]
```

```python
# Fit SARIMAX on the original (non-log) target data
model = SARIMAX(target_train,
                exog=exog_train,
                order=(1, 1, 1),
                seasonal_order=(0, 0, 0, 0),
                enforce_stationarity=False,
                enforce_invertibility=False)
results = model.fit(disp=False)
print(results.summary())
```

```python
# Forecast on the test set - non log
n_steps = len(test_weekly)
forecast = results.get_forecast(steps=n_steps, exog=exog_test).predicted_mean
```

```python
# Compute performance metrics - non log
rmse = np.sqrt(mean_squared_error(target_test, forecast))
mae = mean_absolute_error(target_test, forecast)
epsilon = 1e-10  # small constant to avoid division by zero in MAPE
mape = np.mean(np.abs((target_test - forecast) / (target_test + epsilon))) * 100

print("\nForecast Performance Metrics (Without Log Transformation):")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"Adjusted MAPE: {mape:.2f}%")
```

```python
# Fit SARIMAX on the original (log) target data
model_log = SARIMAX(target_train_log,
                exog=exog_train,
                order=(1, 1, 1),
                seasonal_order=(0, 0, 0, 0),
                enforce_stationarity=False,
                enforce_invertibility=False)
results_log = model_log.fit(disp=False)
print(results_log.summary())
```

```python
# Forecast on the test set
n_steps_log = len(test_weekly)
forecast_log = results_log.get_forecast(steps=n_steps_log, exog=exog_test).predicted_mean
```

```python
# Compute performance metrics
rmse = np.sqrt(mean_squared_error(target_test_log, forecast_log))
mae = mean_absolute_error(target_test_log, forecast_log)
epsilon = 1e-10  # small constant to avoid division by zero in MAPE
mape = np.mean(np.abs((target_test_log - forecast_log) / (target_test_log + epsilon))) * 100

print("\nForecast Performance Metrics (With Log Transformation):")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"Adjusted MAPE: {mape:.2f}%")
```

```python
results_list = []
unique_points = df_wind_daily_final_all_POC_clean['POC_Code'].unique()

for point in unique_points:
    df_point = df_wind_daily_final_all_POC_clean[df_wind_daily_final_all_POC_clean['POC_Code'] == point].copy()
    df_point.sort_index(inplace=True)
    
    # Select only numeric columns for resampling
    numeric_cols = ['Total_TP', 'PS', 'T2M', 'WS50M', 'WD50M']
    df_point_numeric = df_point[numeric_cols]
    
    # Resample to weekly data using mean aggregation
    df_weekly = df_point_numeric.resample('W').mean()
    
    # Ensure we have enough data points
    if len(df_weekly) < 30:
        print(f"Not enough weekly data for POC {point}. Skipping.")
        continue
    
    # Standard 80/20 train-test split based on the number of weeks
    n = len(df_weekly)
    train_size = int(n * 0.8)
    train_weekly = df_weekly.iloc[:train_size]
    test_weekly = df_weekly.iloc[train_size:]
    
    # Define target and exogenous variables
    target_train = train_weekly['Total_TP']
    target_test = test_weekly['Total_TP']
    exog_columns = ['WS50M','PS']
    exog_train = train_weekly[exog_columns]
    exog_test = test_weekly[exog_columns]
    
    # Fit a SARIMAX model (example: ARIMA(1,1,1))
    try:
        model = SARIMAX(target_train,
                        exog=exog_train,
                        order=(1, 1, 1),
                        seasonal_order=(0, 0, 0, 0),
                        enforce_stationarity=False,
                        enforce_invertibility=False)
        results = model.fit(disp=False)
    except Exception as e:
        print(f"Error fitting model for POC {point}: {e}")
        continue
    
    # Forecast for the test period
    n_steps = len(test_weekly)
    try:
        forecast = results.get_forecast(steps=n_steps, exog=exog_test).predicted_mean
    except Exception as e:
        print(f"Error forecasting for POC {point}: {e}")
        continue
    
    rmse = np.sqrt(mean_squared_error(target_test, forecast))
    mae = mean_absolute_error(target_test, forecast)
    epsilon = 1e-10
    mape = np.mean(np.abs((target_test - forecast) / (target_test + epsilon))) * 100
    
    results_list.append({
        'POC_Code': point,
        'Train Weeks': len(train_weekly),
        'Test Weeks': len(test_weekly),
        'RMSE': rmse,
        'MAE': mae,
        'MAPE': mape
    })
    
    print(f"POC {point}: RMSE = {rmse:.2f}, MAE = {mae:.2f}, MAPE = {mape:.2f}%")
    
results_df = pd.DataFrame(results_list)
print("\nForecast Performance by POC:")
print(results_df)
```

```python
# Assuming df_HRP2201_weekly is your weekly resampled DataFrame for HRP2201,
# and that you have already created a log-transformed target column.
df_HRP2201_weekly['Total_TP_log'] = np.log(df_HRP2201_weekly['Total_TP'] + 1)

# Apply first differencing to the log-transformed series
df_HRP2201_weekly['Total_TP_log_diff'] = df_HRP2201_weekly['Total_TP_log'].diff()

# Drop the first row with NaN after differencing
ts_diff = df_HRP2201_weekly['Total_TP_log_diff'].dropna()

# Run ADF test on the differenced series
result_diff = adfuller(ts_diff)
print("ADF Statistic (first difference):", result_diff[0])
print("p-value (first difference):", result_diff[1])
print("Critical Values:")
for key, value in result_diff[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# Then split the data into train and test sets, e.g.:
n = len(df_HRP2201_weekly)
train_size = int(n * 0.8)
train_weekly = df_HRP2201_weekly.iloc[:train_size]
test_weekly = df_HRP2201_weekly.iloc[train_size:]

# Define the target variable for modeling as the differenced log series:
target_train_diff = train_weekly['Total_TP_log_diff']
target_test_diff = test_weekly['Total_TP_log_diff']

# Define exogenous variables as needed (if any). Here, we assume none for simplicity.
exog_cols = ['WS50M','PS']
exog_train = train_weekly[exog_cols]
exog_test = test_weekly[exog_cols]

# exog_train = None
# exog_test = None

# Fit SARIMAX on the differenced log series, with d=0 since we already differenced manually:
model_diff = SARIMAX(target_train_diff,
                     exog=exog_train,
                     order=(1, 1, 1),  # Example order; you'll need to tune this
                     seasonal_order=(0, 0, 0, 0),
                     enforce_stationarity=False,
                     enforce_invertibility=False)
results_diff = model_diff.fit(disp=False)
print(results_diff.summary())

# Forecast the differenced log series over the test period:
n_steps = len(test_weekly)
forecast_diff = results_diff.get_forecast(steps=n_steps, exog=exog_test).predicted_mean

# To recover the forecast on the log scale, you need to add the forecasted differences
# to the last observed log value from the training set, then cumulatively sum.
last_train_log = train_weekly['Total_TP_log'].iloc[-1]
forecast_log = last_train_log + forecast_diff.cumsum()

# Back-transform to original scale:
forecast_original = np.exp(forecast_log) - 1

# Compare with actual Total_TP in test set (back-transform accordingly if needed)
target_test_original = np.exp(test_weekly['Total_TP_log']) - 1

# Compute performance metrics:
from sklearn.metrics import mean_squared_error, mean_absolute_error
rmse = np.sqrt(mean_squared_error(target_test_original, forecast_original))
mae = mean_absolute_error(target_test_original, forecast_original)
epsilon = 1e-10
mape = np.mean(np.abs((target_test_original - forecast_original) / (target_test_original + epsilon))) * 100

print("Forecast Performance Metrics (Differenced Log Model):")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"Adjusted MAPE: {mape:.2f}%")
```
