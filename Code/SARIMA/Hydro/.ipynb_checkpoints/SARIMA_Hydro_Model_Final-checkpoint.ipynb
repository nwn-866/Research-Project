# Forecasting Weekly Hydro Energy Generation in New Zealand
Model - SARIMA

This notebook addresses:
* RQ1: How well does a SARIMA model forecast weekly hydro energy generation?
* RQ2: Does incorporating weekly-aggregated climate variables into SARIMAX improve forecasting accuracy?

The analysis uses weekly sumed and weekly averaged NASA climate data, aligned with operational hydro generation data.

```python
# Add 1 to 2-week lagged climate features
df_hydro_weekly_north['TP_lag1w'] = df_hydro_weekly_north['Total_TP'].shift(1)
df_hydro_weekly_north['T2M_lag1w'] = df_hydro_weekly_north['T2M'].shift(1)
df_hydro_weekly_north['GWETPROF_lag1w'] = df_hydro_weekly_north['GWETPROF'].shift(1)
df_hydro_weekly_north['RH2M_lag1w'] = df_hydro_weekly_north['RH2M'].shift(1)
df_hydro_weekly_north['PRECIP_lag1w'] = df_hydro_weekly_north['PRECTOTCORR'].shift(1)
df_hydro_weekly_north['EVLAND_lag1w'] = df_hydro_weekly_north['EVLAND'].shift(1)
df_hydro_weekly_north['PRECIP_lag2w'] = df_hydro_weekly_north['PRECTOTCORR'].shift(2)
df_hydro_weekly_north['PRECIP_EWMA'] = df_hydro_weekly_north['PRECTOTCORR'].ewm(alpha=0.3, adjust=False).mean()
# EWMA for EVLAND with a “slow” decay (alpha = 0.1)
df_hydro_weekly_north['EVLAND_EWMA'] = df_hydro_weekly_north['EVLAND'].ewm(alpha=0.3, adjust=False).mean()
df_hydro_weekly_north['APIxGWETPROF'] = df_hydro_weekly_north['PRECIP_EWMA'] * df_hydro_weekly_north['GWETPROF']
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -----------------------------------------------------------------------------
# ASSUMPTIONS / PRE‐DEFINED VARIABLES
# -----------------------------------------------------------------------------
# • series_n : pandas.Series of weekly North Island “Total_TP” (kWh), indexed by date.
# -----------------------------------------------------------------------------

# 1) Create train/test split (80/20)
n_n      = len(series_n)
split_n  = int(n_n * 0.8)

train_n = series_n.iloc[:split_n].copy()
test_n  = series_n.iloc[split_n:].copy()

# 2) Define the four candidate (order, seasonal_order) pairs
candidates = [
    ((1, 0, 1), (0, 0, 0, 52)),
    ((0, 0, 1), (0, 0, 1, 52)),
    ((1, 0, 1), (1, 0, 0, 52)),
    ((0, 0, 2), (1, 0, 0, 52)),
    ((1, 0, 0), (0, 0, 1, 52)),
    ((1, 0, 1), (0, 0, 1, 52))
]

# 3) Loop through candidates, fit, forecast, and compute metrics
results_list = []

for idx, ((p, d, q), (P, D, Q, s)) in enumerate(candidates, start=1):
    # 3a) Fit SARIMAX on the TRAIN segment
    model = SARIMAX(
        train_n,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res = model.fit(disp=False)

    # 3b) Forecast on TEST segment
    n_steps = len(test_n)
    fc      = res.get_forecast(steps=n_steps)

    # 3c) Align the forecasted mean with test_n.index
    pred       = pd.Series(fc.predicted_mean.values, index=test_n.index)
    conf_int   = fc.conf_int().copy()
    conf_int.index = test_n.index

    # 3d) Compute MAE, RMSE, MAPE
    y_true = test_n.values
    y_pred = pred.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    # 3e) Collect results
    results_list.append({
        "candidate_index":   idx,
        "order":             (p, d, q),
        "seasonal_order":    (P, D, Q, s),
        "AIC":               res.aic,
        "BIC":               res.bic,
        "MAE":               mae,
        "RMSE":              rmse,
        "MAPE":              mape
    })

# 4) Convert to DataFrame and display
df_results = pd.DataFrame(results_list)
print(df_results.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -----------------------------------------------------------------------------
# ASSUMPTIONS / PREDEFINED:
#
# • weekly_lagged_north : pandas.DataFrame indexed by DATE (weekly), with columns:
#       – 'Total_TP'        (this is the target series)
#       – 'APIxGWETPROF'
#       – 'PRECIP_EWMA'
#       – 'GWETPROF'
#       – 'RH2M_lag1w'
#       – 'EVLAND_lag1w'
#       – 'T2M_lag1w'
#       – … (and any other columns you might have)
#
# • We already know from your ADF/pacf/… work that the best (p,d,q,P,D,Q,s) for North‐island wind is:
p, d, q   = 0, 0, 2        # nonseasonal MA(2)
P, D, Q, s = 1, 0, 0, 52   # single seasonal AR(1) at lag 52
# -----------------------------------------------------------------------------

# 1) Train/test split (80/20) on the *entire* DataFrame
n_n     = len(weekly_lagged_north)
split_n = int(n_n * 0.8)

# The target series is 'Total_TP'
train_y = weekly_lagged_north['Total_TP'].iloc[:split_n].copy()
test_y  = weekly_lagged_north['Total_TP'].iloc[split_n:].copy()

# The SAME split indices will apply to exogenous columns as well.
train_df = weekly_lagged_north.iloc[:split_n].copy()
test_df  = weekly_lagged_north.iloc[split_n:].copy()

# 2) List of all single‐column exogenous candidates
exog_cols = ['TP_lag1w','APIxGWETPROF','PRECIP_EWMA','GWETPROF','RH2M_lag1w']

results_exog = []

for col in exog_cols:
    # 2a) Pull that single column out for train/test
    exog_train = train_df[[col]].copy()
    exog_test  = test_df[[col]].copy()
    
    # 2b) It’s possible the first row or two in exog_train are NaN
    #     (e.g. because RH2M_lag1w or EVLAND_lag1w have to be shifted). 
    #     So we drop any leading NaNs so that indices align:
    first_valid = exog_train[col].first_valid_index()
    if first_valid is not None:
        train_y_trim    = train_y.loc[first_valid:].copy()
        exog_train_trim = exog_train.loc[first_valid:].copy()
    else:
        # If no NaNs, just keep everything
        train_y_trim    = train_y.copy()
        exog_train_trim = exog_train.copy()
    
    # For the test period, we assume lagged exogs are already aligned.
    # If your test exog also has NaNs you can drop those same dates, but
    # typically only the train side has missing rows from lags.
    test_y_trim    = test_y.copy()
    exog_test_trim = exog_test.copy()
    
    # 2c) Fit SARIMAX(train_y_trim, exog=exog_train_trim)
    model_exog = SARIMAX(
        train_y_trim,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        exog=exog_train_trim,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res_exog = model_exog.fit(disp=False)
    
    # 2d) Forecast on the test set, passing exog_test_trim
    n_steps = len(test_y_trim)
    fc      = res_exog.get_forecast(steps=n_steps, exog=exog_test_trim)
    
    # Align the predicted mean to the exact same index as test_y_trim
    pred = pd.Series(fc.predicted_mean.values, index=test_y_trim.index)
    
    # 2e) Compute MAE, RMSE, MAPE
    y_true = test_y_trim.values
    y_pred = pred.values
    
    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100
    
    results_exog.append({
        "exog_col": col,
        "AIC":      res_exog.aic,
        "BIC":      res_exog.bic,
        "MAE":      mae,
        "RMSE":     rmse,
        "MAPE":     mape
    })

# 3) Turn the list of dicts into a DataFrame and sort by MAPE
df_exog_res = pd.DataFrame(results_exog)
df_exog_res = df_exog_res.sort_values("MAPE").reset_index(drop=True)

print(df_exog_res.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from itertools import combinations
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -----------------------------------------------------------------------------
# ASSUMPTIONS / PREDEFINED:
# -----------------------------------------------------------------------------
# • weekly_lagged_north : pandas.DataFrame indexed by DATE (weekly), with columns:
#       – 'Total_TP'        (target series)
#       – 'TP_lag1w'
#       – 'RH2M_lag1w'
#       – 'GWETPROF'
#       – 'PRECIP_EWMA'
#       – 'APIxGWETPROF'
#
# • We have already chosen from earlier diagnostics:
#     p, d, q   = 0, 0, 2        # nonseasonal (an MA(2))
#     P, D, Q, s = 1, 0, 0, 52   # seasonal AR(1) at lag 52
# -----------------------------------------------------------------------------

p, d, q   = 0, 0, 2
P, D, Q, s = 1, 0, 0, 52

# 1) Split 80/20
n_n     = len(weekly_lagged_north)
split_n = int(n_n * 0.8)

train_y = weekly_lagged_north['Total_TP'].iloc[:split_n].copy()
test_y  = weekly_lagged_north['Total_TP'].iloc[split_n:].copy()

# We'll slice the SAME indices for all exog columns.
train_df = weekly_lagged_north.iloc[:split_n].copy()
test_df  = weekly_lagged_north.iloc[split_n:].copy()

# 2) List of exogenous column names (from your DataFrame)
exog_cols = [
    'TP_lag1w',
    'RH2M_lag1w',
    'GWETPROF',
    'PRECIP_EWMA',
    'APIxGWETPROF'
]

# -----------------------------------------------------------------------------
# A) SINGLE exog results (you already have these, but we re‐compute for completeness)
# -----------------------------------------------------------------------------
single_results = []

for col in exog_cols:
    exog_train = train_df[[col]].copy()
    exog_test  = test_df[[col]].copy()

    # Drop the very first NA rows if this lag column has NaNs at the top
    first_valid = exog_train[col].first_valid_index()
    if first_valid is not None:
        train_y_trim    = train_y.loc[first_valid:].copy()
        exog_train_trim = exog_train.loc[first_valid:].copy()
    else:
        train_y_trim    = train_y.copy()
        exog_train_trim = exog_train.copy()

    test_y_trim    = test_y.copy()
    exog_test_trim = exog_test.copy()

    model_exog = SARIMAX(
        train_y_trim,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        exog=exog_train_trim,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res_exog = model_exog.fit(disp=False)

    fc   = res_exog.get_forecast(steps=len(test_y_trim), exog=exog_test_trim)
    pred = pd.Series(fc.predicted_mean.values, index=test_y_trim.index)

    y_true = test_y_trim.values
    y_pred = pred.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    single_results.append({
        "exog_cols": (col,),
        "AIC":       res_exog.aic,
        "BIC":       res_exog.bic,
        "MAE":       mae,
        "RMSE":      rmse,
        "MAPE":      mape
    })

df_single = pd.DataFrame(single_results).sort_values("MAPE").reset_index(drop=True)

print("\n=== Single‐Exogenous Results ===")
print(df_single.to_string(index=False))


# -----------------------------------------------------------------------------
# B) PAIRWISE combinations (two exogs at a time)
# -----------------------------------------------------------------------------
pair_results = []

for combo in combinations(exog_cols, 2):
    exog_train = train_df[list(combo)].copy()
    exog_test  = test_df[list(combo)].copy()

    # Drop leading NaNs (in case any lag columns produce NaN at the top)
    # We drop all rows up to the first index where *both* columns are non‐NaN.
    valid_idx   = exog_train.dropna().index.min()
    if valid_idx is not None:
        train_y_trim    = train_y.loc[valid_idx:].copy()
        exog_train_trim = exog_train.loc[valid_idx:].copy()
    else:
        train_y_trim    = train_y.copy()
        exog_train_trim = exog_train.copy()

    test_y_trim    = test_y.copy()
    exog_test_trim = exog_test.copy()

    model_combo = SARIMAX(
        train_y_trim,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        exog=exog_train_trim,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res_combo = model_combo.fit(disp=False)

    fc_combo = res_combo.get_forecast(steps=len(test_y_trim), exog=exog_test_trim)
    pred_combo = pd.Series(fc_combo.predicted_mean.values, index=test_y_trim.index)

    y_true = test_y_trim.values
    y_pred = pred_combo.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    pair_results.append({
        "exog_cols": combo,
        "AIC":       res_combo.aic,
        "BIC":       res_combo.bic,
        "MAE":       mae,
        "RMSE":      rmse,
        "MAPE":      mape
    })

df_pair = pd.DataFrame(pair_results).sort_values("MAPE").reset_index(drop=True)

print("\n=== Pairwise Exogenous Results ===")
print(df_pair.to_string(index=False))


# -----------------------------------------------------------------------------
# C) OPTIONAL: TRIPLE combinations (three exogs at a time)
# -----------------------------------------------------------------------------
triple_results = []

for combo in combinations(exog_cols, 3):
    exog_train = train_df[list(combo)].copy()
    exog_test  = test_df[list(combo)].copy()

    # Drop leading NaNs if any
    valid_idx = exog_train.dropna().index.min()
    if valid_idx is not None:
        train_y_trim    = train_y.loc[valid_idx:].copy()
        exog_train_trim = exog_train.loc[valid_idx:].copy()
    else:
        train_y_trim    = train_y.copy()
        exog_train_trim = exog_train.copy()

    test_y_trim    = test_y.copy()
    exog_test_trim = exog_test.copy()

    model_triple = SARIMAX(
        train_y_trim,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        exog=exog_train_trim,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res_triple = model_triple.fit(disp=False)

    fc_triple = res_triple.get_forecast(steps=len(test_y_trim), exog=exog_test_trim)
    pred_triple = pd.Series(fc_triple.predicted_mean.values, index=test_y_trim.index)

    y_true = test_y_trim.values
    y_pred = pred_triple.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    triple_results.append({
        "exog_cols": combo,
        "AIC":       res_triple.aic,
        "BIC":       res_triple.bic,
        "MAE":       mae,
        "RMSE":      rmse,
        "MAPE":      mape
    })

df_triple = pd.DataFrame(triple_results).sort_values("MAPE").reset_index(drop=True)

print("\n=== Triple‐Exogenous Results ===")
print(df_triple.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -----------------------------------------------------------------------------
# ASSUMPTIONS / PRE‐DEFINED VARIABLES
# -----------------------------------------------------------------------------
# • series_n : pandas.Series of weekly North Island “Total_TP” (kWh), indexed by date.
# -----------------------------------------------------------------------------

# 1) Create train/test split (80/20)
s_s      = len(series_s)
split_n  = int(s_s * 0.8)

train_s = series_s.iloc[:split_n].copy()
test_s  = series_s.iloc[split_n:].copy()

# 2) Define the four candidate (order, seasonal_order) pairs
candidates = [
    ((1, 0, 1), (0, 0, 0, 52)),
    ((0, 0, 1), (0, 0, 1, 52)),
    ((1, 0, 1), (1, 0, 0, 52)),
    ((0, 0, 2), (1, 0, 0, 52)),
    ((1, 0, 0), (0, 0, 1, 52)),
    ((1, 0, 1), (0, 0, 1, 52))
]

# 3) Loop through candidates, fit, forecast, and compute metrics
results_list = []

for idx, ((p, d, q), (P, D, Q, s)) in enumerate(candidates, start=1):
    # 3a) Fit SARIMAX on the TRAIN segment
    model = SARIMAX(
        train_s,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res = model.fit(disp=False)

    # 3b) Forecast on TEST segment
    n_steps = len(test_s)
    fc      = res.get_forecast(steps=n_steps)

    # 3c) Align the forecasted mean with test_n.index
    pred       = pd.Series(fc.predicted_mean.values, index=test_s.index)
    conf_int   = fc.conf_int().copy()
    conf_int.index = test_s.index

    # 3d) Compute MAE, RMSE, MAPE
    y_true = test_s.values
    y_pred = pred.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    # 3e) Collect results
    results_list.append({
        "candidate_index":   idx,
        "order":             (p, d, q),
        "seasonal_order":    (P, D, Q, s),
        "AIC":               res.aic,
        "BIC":               res.bic,
        "MAE":               mae,
        "RMSE":              rmse,
        "MAPE":              mape
    })

# 4) Convert to DataFrame and display
df_results = pd.DataFrame(results_list)
print(df_results.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from itertools import combinations
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -----------------------------------------------------------------------------
# ASSUMPTIONS / PREDEFINED:
# -----------------------------------------------------------------------------
# • weekly_lagged_north : pandas.DataFrame indexed by DATE (weekly), with columns:
#       – 'Total_TP'        (target series)
#       – 'TP_lag1w'
#       – 'RH2M_lag1w'
#       – 'GWETPROF'
#       – 'PRECIP_EWMA'
#       – 'APIxGWETPROF'
#
# • We have already chosen from earlier diagnostics:
#     p, d, q   = 0, 0, 2        # nonseasonal (an MA(2))
#     P, D, Q, s = 1, 0, 0, 52   # seasonal AR(1) at lag 52
# -----------------------------------------------------------------------------

p, d, q   = 0, 0, 2
P, D, Q, s = 1, 0, 0, 52

# 1) Split 80/20
n_n     = len(weekly_lagged_south)
split_n = int(n_n * 0.8)

train_y = weekly_lagged_south['Total_TP'].iloc[:split_n].copy()
test_y  = weekly_lagged_south['Total_TP'].iloc[split_n:].copy()

# We'll slice the SAME indices for all exog columns.
train_df = weekly_lagged_south.iloc[:split_n].copy()
test_df  = weekly_lagged_south.iloc[split_n:].copy()

# 2) List of exogenous column names (from your DataFrame)
exog_cols = ['TP_lag1w','APIxGWETPROF','PRECIP_EWMA','GWETPROF_lag1w']

# -----------------------------------------------------------------------------
# A) SINGLE exog results (you already have these, but we re‐compute for completeness)
# -----------------------------------------------------------------------------
single_results = []

for col in exog_cols:
    exog_train = train_df[[col]].copy()
    exog_test  = test_df[[col]].copy()

    # Drop the very first NA rows if this lag column has NaNs at the top
    first_valid = exog_train[col].first_valid_index()
    if first_valid is not None:
        train_y_trim    = train_y.loc[first_valid:].copy()
        exog_train_trim = exog_train.loc[first_valid:].copy()
    else:
        train_y_trim    = train_y.copy()
        exog_train_trim = exog_train.copy()

    test_y_trim    = test_y.copy()
    exog_test_trim = exog_test.copy()

    model_exog = SARIMAX(
        train_y_trim,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        exog=exog_train_trim,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res_exog = model_exog.fit(disp=False)

    fc   = res_exog.get_forecast(steps=len(test_y_trim), exog=exog_test_trim)
    pred = pd.Series(fc.predicted_mean.values, index=test_y_trim.index)

    y_true = test_y_trim.values
    y_pred = pred.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    single_results.append({
        "exog_cols": (col,),
        "AIC":       res_exog.aic,
        "BIC":       res_exog.bic,
        "MAE":       mae,
        "RMSE":      rmse,
        "MAPE":      mape
    })

df_single = pd.DataFrame(single_results).sort_values("MAPE").reset_index(drop=True)

print("\n=== Single‐Exogenous Results ===")
print(df_single.to_string(index=False))


# -----------------------------------------------------------------------------
# B) PAIRWISE combinations (two exogs at a time)
# -----------------------------------------------------------------------------
pair_results = []

for combo in combinations(exog_cols, 2):
    exog_train = train_df[list(combo)].copy()
    exog_test  = test_df[list(combo)].copy()

    # Drop leading NaNs (in case any lag columns produce NaN at the top)
    # We drop all rows up to the first index where *both* columns are non‐NaN.
    valid_idx   = exog_train.dropna().index.min()
    if valid_idx is not None:
        train_y_trim    = train_y.loc[valid_idx:].copy()
        exog_train_trim = exog_train.loc[valid_idx:].copy()
    else:
        train_y_trim    = train_y.copy()
        exog_train_trim = exog_train.copy()

    test_y_trim    = test_y.copy()
    exog_test_trim = exog_test.copy()

    model_combo = SARIMAX(
        train_y_trim,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        exog=exog_train_trim,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res_combo = model_combo.fit(disp=False)

    fc_combo = res_combo.get_forecast(steps=len(test_y_trim), exog=exog_test_trim)
    pred_combo = pd.Series(fc_combo.predicted_mean.values, index=test_y_trim.index)

    y_true = test_y_trim.values
    y_pred = pred_combo.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    pair_results.append({
        "exog_cols": combo,
        "AIC":       res_combo.aic,
        "BIC":       res_combo.bic,
        "MAE":       mae,
        "RMSE":      rmse,
        "MAPE":      mape
    })

df_pair = pd.DataFrame(pair_results).sort_values("MAPE").reset_index(drop=True)

print("\n=== Pairwise Exogenous Results ===")
print(df_pair.to_string(index=False))


# -----------------------------------------------------------------------------
# C) OPTIONAL: TRIPLE combinations (three exogs at a time)
# -----------------------------------------------------------------------------
triple_results = []

for combo in combinations(exog_cols, 3):
    exog_train = train_df[list(combo)].copy()
    exog_test  = test_df[list(combo)].copy()

    # Drop leading NaNs if any
    valid_idx = exog_train.dropna().index.min()
    if valid_idx is not None:
        train_y_trim    = train_y.loc[valid_idx:].copy()
        exog_train_trim = exog_train.loc[valid_idx:].copy()
    else:
        train_y_trim    = train_y.copy()
        exog_train_trim = exog_train.copy()

    test_y_trim    = test_y.copy()
    exog_test_trim = exog_test.copy()

    model_triple = SARIMAX(
        train_y_trim,
        order=(p, d, q),
        seasonal_order=(P, D, Q, s),
        exog=exog_train_trim,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res_triple = model_triple.fit(disp=False)

    fc_triple = res_triple.get_forecast(steps=len(test_y_trim), exog=exog_test_trim)
    pred_triple = pd.Series(fc_triple.predicted_mean.values, index=test_y_trim.index)

    y_true = test_y_trim.values
    y_pred = pred_triple.values

    mae   = mean_absolute_error(y_true, y_pred)
    rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
    mape  = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-10))) * 100

    triple_results.append({
        "exog_cols": combo,
        "AIC":       res_triple.aic,
        "BIC":       res_triple.bic,
        "MAE":       mae,
        "RMSE":      rmse,
        "MAPE":      mape
    })

df_triple = pd.DataFrame(triple_results).sort_values("MAPE").reset_index(drop=True)

print("\n=== Triple‐Exogenous Results ===")
print(df_triple.to_string(index=False))
```
