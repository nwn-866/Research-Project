```python
# Get list of POC codes to load
selected_pocs = nsp_info_Hydro['POC_Code'].unique()
```

```python
selected_pocs
```

```python
# Initialize empty list to store DataFrames
dfs = []
```

```python
# Loop through each selected POC
for poc in selected_pocs:
    file_EVLAND = os.path.join(
        Input_folder_EVLAND, f"NASA_{poc}_EVLAND_20150101_20250228.csv"
    )
    
    # Check if file exists before loading
    if os.path.exists(file_EVLAND):
        # Skip metadata rows and load actual data
        print(f"Loading file: {file_EVLAND}")
        with open(file_EVLAND, 'r') as f:
            for i, line in enumerate(f):
                if line.startswith('YEAR') or 'DATE' in line:
                    header_line = i
                    break
        #df = pd.read_csv(file_path, skiprows=10)
        df_EVLAND = pd.read_csv(file_EVLAND, skiprows=header_line)
        
        # Add identifier column
        df_EVLAND['POC_Code'] = poc
        
        dfs.append(df_EVLAND)
    else:
        print(f"File not found: {file_path}")

# Combine all into a single DataFrame
combined_Hydro_data_EVLAND = pd.concat(dfs, ignore_index=True)
```

```python
combined_Hydro_data_EVLAND
```

```python
# Create a new date column
combined_Hydro_data_EVLAND['DATE'] = pd.to_datetime(
    combined_Hydro_data_EVLAND['YEAR'].astype(str) + combined_Hydro_data_EVLAND['DOY'].astype(str),
    format='%Y%j'
)

```

```python
# Optional: Set DATE as index or reorder columns
# combined_Hydro_data.set_index('DATE', inplace=True)

# Preview the updated DataFrame
print(combined_Hydro_data_EVLAND[['DATE', 'YEAR', 'DOY', 'EVLAND', 'POC_Code']].head())
```

```python
combined_Hydro_data_EVLAND
```

```python
# 2. Filter data between 2015-01-01 and 2024-12-31
filtered_Hydro_data_EVLAND = combined_Hydro_data_EVLAND[
    (combined_Hydro_data_EVLAND['DATE'] >= '2015-01-01') &
    (combined_Hydro_data_EVLAND['DATE'] <= '2024-12-31')
]
```

```python
filtered_Hydro_data_EVLAND = filtered_Hydro_data_EVLAND[['DATE', 'POC_Code', 'EVLAND']]
```

```python
daily_avg_EVLAND = filtered_Hydro_data_EVLAND.groupby('DATE', as_index=False)['EVLAND'].mean()
```
