```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.callbacks import EarlyStopping
import random
```

```python

# 1: Load all datasets 
wind_data = pd.read_csv('wind_data.csv')
avg_ps = pd.read_csv('1Wind_Factor_Daily_AVG_PS.csv')
avg_t2m = pd.read_csv('2Wind_Factor_Daily_AVG_T2M.csv')
avg_wd50m = pd.read_csv('3Wind_Factor_Daily_AVG_WD50M.csv')
avg_ws50m = pd.read_csv('4Wind_Factor_Daily_AVG_WS50M_Formatted.csv')

# 2: Convert DATE columns to datetime (fixing format issues) 
wind_data['Trading_date'] = pd.to_datetime(wind_data['Trading_date'])
avg_ps['DATE'] = pd.to_datetime(avg_ps['DATE'])
avg_t2m['DATE'] = pd.to_datetime(avg_t2m['DATE'], dayfirst=True) 
avg_wd50m['DATE'] = pd.to_datetime(avg_wd50m['DATE'])
avg_ws50m['DATE'] = pd.to_datetime(avg_ws50m['DATE'])

# Rename Trading_date to DATE for merging
wind_data = wind_data.rename(columns={'Trading_date': 'DATE'})

wind_data
```

```python
# 3: Merge all datasets on DATE 
merged_df = wind_data[['DATE', 'Total_TP']] \
    .merge(avg_ps, on='DATE') \
    .merge(avg_t2m, on='DATE') \
    .merge(avg_wd50m, on='DATE') \
    .merge(avg_ws50m, on='DATE')

#  4: Set DATE as index 
merged_df.set_index('DATE', inplace=True)

# 5: Resample to weekly frequency (mean per week) 
# Note: resampling handles leap years automatically
weekly_df = merged_df.resample('W').mean()

#  Check result 
print("Weekly data shape:", weekly_df.shape)
print(weekly_df.head())
```

```python
#   Set seeds for reproducibility  
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

#   Feature Engineering  
weekly_df['WS50M_lag1'] = weekly_df['Avg_WS50M'].shift(1)
weekly_df['WD50M_lag1'] = weekly_df['Avg_WD50M'].shift(1)
weekly_df['Total_TP_lag1'] = weekly_df['Total_TP'].shift(1)
weekly_df['Total_TP_diff'] = weekly_df['Total_TP'].diff()
weekly_df['WS50M_diff'] = weekly_df['Avg_WS50M'].diff()
weekly_df['WD50M_rolling3'] = weekly_df['Avg_WD50M'].rolling(3).mean()
weekly_df.dropna(inplace=True)

#   Feature Selection  
features = ['Avg_PS', 'Avg_T2M', 'Avg_WD50M', 'Avg_WS50M',
            'WS50M_lag1', 'WD50M_lag1', 'Total_TP_lag1',
            'Total_TP_diff', 'WS50M_diff', 'WD50M_rolling3']

X = weekly_df[features]
y = weekly_df['Total_TP']

#   Scaling  
X_scaler = MinMaxScaler()
X_scaled = X_scaler.fit_transform(X)

y_scaler = MinMaxScaler()
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))

#   Train-test split  
train_index = weekly_df.index < '2024-01-01'
X_train, X_test = X_scaled[train_index], X_scaled[~train_index]
y_train, y_test = y_scaled[train_index], y_scaled[~train_index]

#   ANN Model  
model = Sequential([
    Dense(128, input_shape=(X_train.shape[1],)),
    LeakyReLU(),
    BatchNormalization(),
    Dense(64),
    LeakyReLU(),
    BatchNormalization(),
    Dense(32),
    LeakyReLU(),
    Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")

#   Early stopping  
early_stop = EarlyStopping(patience=15, restore_best_weights=True)

#   Train  
history = model.fit(X_train, y_train,
                    epochs=250,
                    batch_size=16,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stop],
                    verbose=1)

#   Predict  
y_pred_scaled = model.predict(X_test)
y_pred_rescaled = y_scaler.inverse_transform(y_pred_scaled)
y_test_rescaled = y_scaler.inverse_transform(y_test)

#   Evaluate  
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))

print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")

#   Plot Actual vs Predicted  
plt.figure(figsize=(14, 6))
plt.plot(weekly_df.index[~train_index], y_test_rescaled, label="Actual Wind Energy")
plt.plot(weekly_df.index[~train_index], y_pred_rescaled, label="Predicted Wind Energy", linestyle="--", color="red")
plt.title("Actual vs Predicted Wind Energy Generation (2024) with Enhanced ANN (Reproducible)")
plt.xlabel("Week")
plt.ylabel("Total_TP (Wind Energy)")
plt.legend()
plt.tight_layout()
plt.show()
```
