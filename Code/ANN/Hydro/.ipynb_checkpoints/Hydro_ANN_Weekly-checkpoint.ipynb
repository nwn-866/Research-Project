```python
# Create a new DataFrame excluding the 'Island' column
df_island_north_new = df_island_north.drop(columns=['Island'])
```

```python
df_island_north_new
```

```python
# 5: Resample to weekly frequency (mean per week) 
# Note: resampling handles leap years automatically
df_island_north_weekly = df_island_north_new.resample('W').mean()

#  Check result 
print("Weekly data shape:", df_island_north_weekly.shape)
print(df_island_north_weekly.head())
```

```python
#   Set seeds for reproducibility  
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

#   Feature Engineering  
df_island_north_weekly['WS50M_lag1'] = df_island_north_weekly['WS50M'].shift(1)
df_island_north_weekly['WD50M_lag1'] = df_island_north_weekly['WD50M'].shift(1)
df_island_north_weekly['Total_TP_lag1'] = df_island_north_weekly['Total_TP'].shift(1)
df_island_north_weekly['Total_TP_diff'] = df_island_north_weekly['Total_TP'].diff()
df_island_north_weekly['WS50M_diff'] = df_island_north_weekly['WS50M'].diff()
df_island_north_weekly['WD50M_rolling3'] = df_island_north_weekly['WD50M'].rolling(3).mean()
df_island_north_weekly.dropna(inplace=True)

```

```python
# #   Feature Selection  
# features = ['PS', 'T2M', 'WD50M', 'WS50M',
#             'WS50M_lag1', 'WD50M_lag1', 'Total_TP_lag1',
#             'Total_TP_diff', 'WS50M_diff', 'WD50M_rolling3']

```

```python
#   Feature Selection  
features = ['WS50M',
            'WS50M_lag1', 'Total_TP_lag1',
            'Total_TP_diff', 'WS50M_diff']
```

```python
# #   Feature Selection  
# features = ['PS', 'WS50M','WS50M_lag1', 'Total_TP_lag1','Total_TP_diff', 'WS50M_diff']
```

```python
X = df_island_north_weekly[features]
y = df_island_north_weekly['Total_TP']
```

```python
#   Scaling  
X_scaler = MinMaxScaler()
X_scaled = X_scaler.fit_transform(X)

y_scaler = MinMaxScaler()
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))

```

```python
#   Train-test split  
train_index = df_island_north_weekly.index < '2024-01-01'
X_train, X_test = X_scaled[train_index], X_scaled[~train_index]
y_train, y_test = y_scaled[train_index], y_scaled[~train_index]
```

```python
# # Recreate the mask using the same DataFrame as X_scaled
# train_index = df_island_north_numeric.index < '2024-01-01'

# # Apply the split using correct index
# X_train, X_test = X_scaled[train_index], X_scaled[~train_index]
# y_train, y_test = y_scaled[train_index], y_scaled[~train_index]
```

```python
#   ANN Model  
model = Sequential([
    Dense(128, input_shape=(X_train.shape[1],)),
    LeakyReLU(),
    BatchNormalization(),
    Dense(64),
    LeakyReLU(),
    BatchNormalization(),
    Dense(32),
    LeakyReLU(),
    Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")

```

```python
#   Early stopping  
early_stop = EarlyStopping(patience=15, restore_best_weights=True)

```

```python
#   Train  
history = model.fit(X_train, y_train,
                    epochs=250,
                    batch_size=16,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stop],
                    verbose=1)

```

```python
#   Predict  
y_pred_scaled = model.predict(X_test)
y_pred_rescaled = y_scaler.inverse_transform(y_pred_scaled)
y_test_rescaled = y_scaler.inverse_transform(y_test)

```

```python
#   Evaluate  
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))

print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")

```

```python
# MAE: 73,011.87
# RMSE: 94,425.64
# MAPE: 9.32%
```

```python
epsilon = 1e-10  # to prevent division by zero
mape = np.mean(np.abs((y_test_rescaled - y_pred_rescaled) / (y_test_rescaled + epsilon))) * 100

print(f"MAPE: {mape:.2f}%")
```

```python
#   Plot Actual vs Predicted  
plt.figure(figsize=(14, 6))
plt.plot(df_island_north_weekly.index[~train_index], y_test_rescaled, label="Actual Wind Energy")
plt.plot(df_island_north_weekly.index[~train_index], y_pred_rescaled, label="Predicted Wind Energy - North Island (Weekly)", linestyle="--", color="red")
plt.title("Actual vs Predicted Wind Energy Generation (2024) with Enhanced ANN (Reproducible)")
plt.xlabel("Weekly")
plt.ylabel("Total_TP (Wind Energy)")
plt.legend()
plt.tight_layout()
plt.show()
```
