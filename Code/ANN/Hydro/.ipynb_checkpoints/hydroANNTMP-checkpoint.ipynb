```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Set Random Seed for Reproducibility 
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
import pandas as pd

#   Step 2: Load Datasets  
hydro_data = pd.read_csv('hydro_data.csv')
evland = pd.read_csv('1Hydro_Factor_Daily_AVG_EVLAND.csv')
rh2m = pd.read_csv('2Hydro_Factor_Daily_AVG_RH2M.csv')
prectot = pd.read_csv('3Hydro_Factor_Daily_TOT_PRECTOTCORR.csv')
t2m = pd.read_csv('4Wind_Factor_Daily_AVG_T2M.csv')

#   Step 3: Convert Dates to datetime format  
hydro_data['Trading_date'] = pd.to_datetime(hydro_data['Trading_date'])
evland['DATE'] = pd.to_datetime(evland['DATE'])
rh2m['DATE'] = pd.to_datetime(rh2m['DATE'])
prectot['DATE'] = pd.to_datetime(prectot['DATE'])
t2m['DATE'] = pd.to_datetime(t2m['DATE'])
```

```python
#   Step 4: Rename hydro_data's date column for merging  
hydro_data = hydro_data.rename(columns={'Trading_date': 'DATE'})
```

```python
#   Step 5: Merge all datasets on DATE  
merged_df = hydro_data[['DATE', 'Total_TP']] \
    .merge(evland, on='DATE') \
    .merge(rh2m, on='DATE') \
    .merge(prectot, on='DATE') \
    .merge(t2m, on='DATE')

#   Step 6: Set DATE as index  
merged_df.set_index('DATE', inplace=True)

#   Step 7: Resample to Weekly Frequency (Mean per Week)  
weekly_df = merged_df.resample('W').mean()

#   Step 8: Check Results  
print("Weekly data shape:", weekly_df.shape)
print(weekly_df.head(10))
```

```python
#  Feature Engineering 
weekly_df['Total_TP_lag1'] = weekly_df['Total_TP'].shift(1)
weekly_df['Total_TP_lag2'] = weekly_df['Total_TP'].shift(2)
weekly_df['Total_TP_rolling3'] = weekly_df['Total_TP'].rolling(3).mean()
weekly_df['T2M_RH2M_interaction'] = weekly_df['Avg_T2M'] * weekly_df['AVG_RH2M']
weekly_df.dropna(inplace=True)
```

```python
#  Features & Target 
features = [
    'Avg_EVLAND', 'AVG_RH2M', 'Tot_PRECTOTCORR', 'Avg_T2M',
    'Total_TP_lag1', 'Total_TP_lag2', 'Total_TP_rolling3',
    'T2M_RH2M_interaction'
]
X = weekly_df[features]
y = weekly_df['Total_TP']
```

```python

#  Normalize 
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=features, index=X.index)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))

#  Train-Test Split 
train_index = X_scaled.index < '2024-01-01'
X_train, X_test = X_scaled[train_index], X_scaled[~train_index]
y_train, y_test = y_scaled[train_index], y_scaled[~train_index]
```

```python
#  ANN Model 
model = Sequential([
    Dense(256, input_shape=(X_train.shape[1],)), LeakyReLU(), BatchNormalization(), Dropout(0.2),
    Dense(128), LeakyReLU(), BatchNormalization(), Dropout(0.2),
    Dense(64), LeakyReLU(),
    Dense(1)
])
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005), loss='mse')

#  Training 
early_stop = EarlyStopping(patience=25, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=300, batch_size=16,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stop], verbose=0)
```

```python
#  Prediction & Evaluation 
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_test_actual = scaler_y.inverse_transform(y_test)

mae = mean_absolute_error(y_test_actual, y_pred)
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred))
mape = np.mean(np.abs((y_test_actual - y_pred) / y_test_actual)) * 100
```

```python
#  Plot 
plt.figure(figsize=(14, 6))
plt.plot(y_test_actual, label='Actual', linewidth=2)
plt.plot(y_pred, label='Predicted', linestyle='--', color='red')
plt.title('Hydro Renewable Energy Forecast vs Actual (ANN Model)\nWith Strategic Lag & Rolling Terms')
plt.xlabel('Date')
plt.ylabel('Total_TP (Hydro Energy)')
plt.legend()
plt.tight_layout()
plt.show()

```

```python
#  Metrics
print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"MAPE: {mape:.2f}%")
```
