```python
#   Set seeds for reproducibility  
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

#   Feature Engineering  
df_weekly_hydro_island_north['PRECTOTCORR_lag1'] = df_weekly_hydro_island_north['PRECTOTCORR'].shift(1)
#df_weekly_hydro_island_north['PRECTOTCORR_lag2'] = df_weekly_hydro_island_north['PRECTOTCORR'].shift(2)
df_weekly_hydro_island_north['Total_TP_lag1'] = df_weekly_hydro_island_north['Total_TP'].shift(1)
df_weekly_hydro_island_north['Total_TP_lag2'] = df_weekly_hydro_island_north['Total_TP'].shift(2)
df_weekly_hydro_island_north['Total_TP_diff'] = df_weekly_hydro_island_north['Total_TP'].diff()
# df_island_north['WS50M_diff'] = df_island_north['WS50M'].diff()
df_weekly_hydro_island_north['Total_TP_rolling3'] = df_weekly_hydro_island_north['Total_TP'].rolling(3).mean()
df_weekly_hydro_island_north.dropna(inplace=True)

# weekly_df['Total_TP_lag1'] = weekly_df['Total_TP'].shift(1)
# weekly_df['Total_TP_lag2'] = weekly_df['Total_TP'].shift(2)
# weekly_df['Total_TP_rolling3'] = weekly_df['Total_TP'].rolling(3).mean()
# weekly_df['T2M_RH2M_interaction'] = weekly_df['Avg_T2M'] * weekly_df['AVG_RH2M']
```

```python
print(df_weekly_hydro_island_north.columns)
print(df_weekly_hydro_island_north.head())
```

```python
# #   Feature Selection  
# features = ['PS', 'T2M', 'WD50M', 'WS50M',
#             'WS50M_lag1', 'WD50M_lag1', 'Total_TP_lag1',
#             'Total_TP_diff', 'WS50M_diff', 'WD50M_rolling3']

```

```python
#   Feature Selection  
features = ['PRECTOTCORR', 'EVLAND', 'RH2M', 'GWETPROF', 'T2M','PRECTOTCORR_lag1', 'Total_TP_lag1', 'Total_TP_lag2','Total_TP_diff','Total_TP_rolling3']
```

```python
X = df_weekly_hydro_island_north[features]
y = df_weekly_hydro_island_north['Total_TP']
```

```python
#   Scaling  
X_scaler = MinMaxScaler()
X_scaled = X_scaler.fit_transform(X)

y_scaler = MinMaxScaler()
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))

```

```python
#   Train-test split  
train_index = df_weekly_hydro_island_north.index < '2024-01-01'
X_train, X_test = X_scaled[train_index], X_scaled[~train_index]
y_train, y_test = y_scaled[train_index], y_scaled[~train_index]
```

```python
#   ANN Model  
model = Sequential([
    Dense(128, input_shape=(X_train.shape[1],)),
    LeakyReLU(),
    BatchNormalization(),
    Dense(64),
    LeakyReLU(),
    BatchNormalization(),
    Dense(32),
    LeakyReLU(),
    Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")

```

```python
#   Early stopping  
early_stop = EarlyStopping(patience=15, restore_best_weights=True)

```

```python
#   Train  
history = model.fit(X_train, y_train,
                    epochs=250,
                    batch_size=16,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stop],
                    verbose=1)

```

```python
#   Predict  
y_pred_scaled = model.predict(X_test)
y_pred_rescaled = y_scaler.inverse_transform(y_pred_scaled)
y_test_rescaled = y_scaler.inverse_transform(y_test)

```

```python
#   Evaluate  
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))

print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")

```

```python
epsilon = 1e-10  # to prevent division by zero
mape = np.mean(np.abs((y_test_rescaled - y_pred_rescaled) / (y_test_rescaled + epsilon))) * 100

print(f"MAPE: {mape:.2f}%")
```

```python
#   Plot Actual vs Predicted  
plt.figure(figsize=(14, 6))
plt.plot(df_weekly_hydro_island_north.index[~train_index], y_test_rescaled, label="Actual Hydro Energy")
plt.plot(df_weekly_hydro_island_north.index[~train_index], y_pred_rescaled, label="Predicted Hydro Energy - North Island (Weekly)", linestyle="--", color="red")
plt.title("Actual vs Predicted hydro Energy Generation (2024) with Enhanced ANN (Reproducible)")
plt.xlabel("Weekly")
plt.ylabel("Total_TP (Hydro Energy)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# df_island_ALL_V = df_island_daily.copy()

# # Group by Island and Date, and compute mean of numeric columns
# df_island_ALL = df_island_ALL_V.groupby(['DATE']).mean().reset_index()

# # Optional: sort by Island and Date
# df_island_ALL = df_island_ALL.sort_values(by=['DATE'])

# # set index to date 
# df_island_ALL.set_index('DATE', inplace=True)

# #   Set seeds for reproducibility  
# np.random.seed(42)
# random.seed(42)
# tf.random.set_seed(42)

# #   Feature Engineering  
# df_weekly_hydro_final_all['PRECTOTCORR_lag1'] = df_weekly_hydro_final_all['PRECTOTCORR'].shift(1)
# df_weekly_hydro_final_all['PRECTOTCORR_lag2'] = df_weekly_hydro_final_all['PRECTOTCORR'].shift(2)
# df_weekly_hydro_final_all['Total_TP_lag1'] = df_weekly_hydro_final_all['Total_TP'].shift(1)
# df_weekly_hydro_final_all['Total_TP_lag2'] = df_weekly_hydro_final_all['Total_TP'].shift(2)
# df_weekly_hydro_final_all['Total_TP_diff'] = df_weekly_hydro_final_all['Total_TP'].diff()
# # df_island_north['WS50M_diff'] = df_island_north['WS50M'].diff()
# df_weekly_hydro_final_all['Total_TP_rolling3'] = df_weekly_hydro_final_all['Total_TP'].rolling(3).mean()
# df_weekly_hydro_final_all.dropna(inplace=True)



# #   Feature Selection  
# features = ['PRECTOTCORR', 'EVLAND', 'RH2M', 'GWETPROF', 'T2M',
#        'PRECTOTCORR_lag1', 'Total_TP_lag1', 'Total_TP_diff',
#        'PRECTOTCORR_lag2', 'Total_TP_lag2'#, 'Total_TP_rolling3'
#            ]


X = df_weekly_hydro_final_all[features]
y = df_weekly_hydro_final_all['Total_TP']

#   Scaling  
X_scaler = MinMaxScaler()
X_scaled = X_scaler.fit_transform(X)

y_scaler = MinMaxScaler()
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))



#   Train-test split  
train_index = df_weekly_hydro_final_all.index < '2024-01-01'
X_train, X_test = X_scaled[train_index], X_scaled[~train_index]
y_train, y_test = y_scaled[train_index], y_scaled[~train_index]


#   ANN Model  
model = Sequential([
    Dense(128, input_shape=(X_train.shape[1],)),
    LeakyReLU(),
    BatchNormalization(),
    Dense(64),
    LeakyReLU(),
    BatchNormalization(),
    Dense(32),
    LeakyReLU(),
    Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")



#   Early stopping  
early_stop = EarlyStopping(patience=15, restore_best_weights=True)



#   Train  
history = model.fit(X_train, y_train,
                    epochs=250,
                    batch_size=16,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stop],
                    verbose=1)

#   Predict  
y_pred_scaled = model.predict(X_test)
y_pred_rescaled = y_scaler.inverse_transform(y_pred_scaled)
y_test_rescaled = y_scaler.inverse_transform(y_test)



#   Evaluate  
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))

print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")



epsilon = 1e-10  # to prevent division by zero
mape = np.mean(np.abs((y_test_rescaled - y_pred_rescaled) / (y_test_rescaled + epsilon))) * 100

print(f"MAPE: {mape:.2f}%")

#   Plot Actual vs Predicted  
plt.figure(figsize=(14, 6))
plt.plot(df_weekly_hydro_final_all.index[~train_index], y_test_rescaled, label="Actual Hydro Energy")
plt.plot(df_weekly_hydro_final_all.index[~train_index], y_pred_rescaled, label="Predicted Hydro Energy - New Zealand (Weekly)", linestyle="--", color="red")
plt.title("Actual vs Predicted Hydro Energy Generation (2024) with Enhanced ANN (Reproducible)")
plt.xlabel("Weekly")
plt.ylabel("Total_TP (Hydro Energy)")
plt.legend()
plt.tight_layout()
plt.show()
```
