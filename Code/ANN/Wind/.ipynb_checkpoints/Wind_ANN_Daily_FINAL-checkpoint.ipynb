# Forecasting Weekly Wind Energy Generation in New Zealand
Model - ANN

This notebook addresses:
* RQ1: How well does a ANN model forecast weekly wind energy generation?
* RQ2: Does incorporating weekly-aggregated climate variables into ANN improve forecasting accuracy?

The analysis uses weekly sumed and weekly averaged NASA climate data, aligned with operational wind generation data.

```python
daily_feats = [
    # 'TP_lag1',        # yesterday’s output
    # 'TP_lag7',        # last week’s output
    # 'GWETPROF',       # ground wetness
    # 'RH2M_mean_7d',   # 7‑day mean humidity
    # 'sin_doy',        # seasonal cycle (day‑of‑year sine)
    # 'PRECIP_7d'       # 7‑day total rainfall

'Total_TP','Total_TP_diff',
'TP_lag1',
'TP_lag7',
#'cos_doy',
#'PS',
#'T2M',
'WS50M'
]
```

```python
X_raw = df_wind_daily_north[daily_feats].values                  # (n_samples, n_features)
y_raw = df_wind_daily_north['Total_TP'].values.reshape(-1, 1) # (n_samples, 1)
```

```python
# 3) Split *once* (no shuffling)
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
    X_raw, y_raw, train_size=0.8, shuffle=False
)
```

```python
# 4) Scale
x_scaler = StandardScaler().fit(X_train_raw)
y_scaler = StandardScaler().fit(y_train_raw)

X_train = x_scaler.transform(X_train_raw)
X_test  = x_scaler.transform(X_test_raw)
y_train = y_scaler.transform(y_train_raw)
y_test  = y_scaler.transform(y_test_raw)
```

```python
# 5) Sanity‐check shapes
print(X_train.shape, y_train.shape)
print(X_test.shape,  y_test.shape)
```

```python
# optimizer with clipping & lower LR
opt = tf.keras.optimizers.Adam(learning_rate=5e-4, clipnorm=1.0)
```

```python
#   ANN Model  
model = Sequential([
    Input(shape=(X_train.shape[1],)),  # Use the Input layer for the input shape
    Dense(128),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(0.2),
    
    Dense(64),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(0.2),
    
    Dense(32),
    LeakyReLU(),
    Dense(1)
])

#model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")

```

```python
model.compile(
    optimizer=opt,
    loss=Huber(delta=0.5),
    metrics=[tf.keras.metrics.MeanAbsolutePercentageError(name='mape')]
)

callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
]
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=200, batch_size=32,
    callbacks=callbacks, verbose=1
)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt

# 1) Predict & inverse‑scale
y_pred_s = model.predict(X_test)
y_pred = y_scaler.inverse_transform(y_pred_s).ravel()
y_true = y_scaler.inverse_transform(y_test).ravel()

```

```python
# 2) Metrics
mae  = mean_absolute_error(y_true, y_pred)
#rmse = mean_squared_error(y_true, y_pred, squared=False)
rmse = root_mean_squared_error(y_true, y_pred)
mape = mean_absolute_percentage_error(y_true, y_pred) * 100
print(f"MAE : {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"MAPE: {mape:.2f}%")

# # 3) Plot
# plt.figure(figsize=(12,5))
# plt.plot(dates_test, y_true, label="Actual")
# plt.plot(dates_test, y_pred, linestyle="--", color="red", label="Predicted")
# plt.title("Sliding‑Window MLP: Actual vs. Predicted")
# plt.xlabel("Date")
# plt.ylabel("Total_TP")
# plt.legend()
# plt.tight_layout()
# plt.show()
```

```python
# 1) Rebuild the “clean” DataFrame exactly as you did before splitting:
df_clean = df_wind_daily_north.copy()
# … your feature engineering here …
df_clean.dropna(inplace=True)

# 2) Compute train_size the same way you did for X_train/X_test:
n = len(df_clean)
train_size = int(0.8 * n)

# 3) Grab the dates for the test slice:
test_dates = df_clean.index[train_size:]

# 4) Now plot, flattening your 1‑D arrays:
plt.figure(figsize=(14, 6))
plt.plot(test_dates, y_true.ravel(), label="Actual Hydro Energy")
plt.plot(test_dates, y_pred.ravel(),       label="Predicted Hydro Energy", linestyle="--", color="red")

plt.title("Actual vs Predicted Hydro Energy Generation")
plt.xlabel("Date")
plt.ylabel("Total_TP (Hydro Energy)")
plt.legend()
plt.tight_layout()
plt.show()
```
