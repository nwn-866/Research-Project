```python
# Get list of POC codes to load
selected_pocs = nsp_info_Wind['POC_Code'].unique()
```

```python
selected_pocs
```

```python
# Initialize empty list to store DataFrames
dfs = []
```

```python
# Loop through each selected POC
for poc in selected_pocs:
    file_T2M = os.path.join(
        Input_folder_T2M, f"NASA_{poc}_T2M_20150101_20250228.csv"
    )
    
    # Check if file exists before loading
    if os.path.exists(file_T2M):
        # Skip metadata rows and load actual data
        print(f"Loading file: {file_T2M}")
        with open(file_T2M, 'r') as f:
            for i, line in enumerate(f):
                if line.startswith('YEAR') or 'DATE' in line:
                    header_line = i
                    break
        #df = pd.read_csv(file_path, skiprows=10)
        df_T2M = pd.read_csv(file_T2M, skiprows=header_line)
        
        # Add identifier column
        df_T2M['POC_Code'] = poc
        
        dfs.append(df_T2M)
    else:
        print(f"File not found: {file_path}")

# Combine all into a single DataFrame
combined_wind_data_T2M = pd.concat(dfs, ignore_index=True)
```

```python
combined_wind_data_T2M
```

```python
# Create a new date column
combined_wind_data_T2M['DATE'] = pd.to_datetime(
    combined_wind_data_T2M['YEAR'].astype(str) + combined_wind_data_T2M['DOY'].astype(str),
    format='%Y%j'
)

```
