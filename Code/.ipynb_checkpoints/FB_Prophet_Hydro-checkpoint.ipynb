```python
#  Step 1: Imports 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  Step 2: Load dataset 
df = pd.read_csv("hydroformatted_dataset.csv", parse_dates=["DATE"])

#  Step 3: Set date index 
df.set_index("DATE", inplace=True)

#  Step 4: Resample Weekly (Sunday start) 
weekly_df = pd.DataFrame()

weekly_df['Total_TP'] = df['Total_TP'].resample('W-SUN').sum()
weekly_df['PRECTOTCORR'] = df['PRECTOTCORR'].resample('W-SUN').sum()

weekly_df['EVLAND'] = df['EVLAND'].resample('W-SUN').mean()
weekly_df['RH2M'] = df['RH2M'].resample('W-SUN').mean()
weekly_df['GWETPROF'] = df['GWETPROF'].resample('W-SUN').mean()
weekly_df['T2M'] = df['T2M'].resample('W-SUN').mean()

#  Step 5: Filter from 2015-03-01 to 2024-12-31 
weekly_df = weekly_df.loc["2015-03-01":"2024-12-31"]

#  Step 6: Reset index for Prophet usage 
# If you don’t name them exactly like ds and y, Prophet will throw an error or fail to understand what to forecast.

weekly_df = weekly_df.reset_index().rename(columns={"DATE": "ds", "Total_TP": "y"})

#  Step 7: Save weekly aggregated file (optional) 
weekly_df.to_csv("weekly_hydro_data.csv", index=False)
```

```python
# Line Plot of Weekly Hydro Energy
plt.figure(figsize=(12, 5))
plt.plot(weekly_df['ds'], weekly_df['y'], color='blue')
plt.title("Weekly Hydro Renewable Energy With Time Duration")
plt.xlabel("Date")
plt.ylabel("Total_TP")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Moving Average (trend insight)
weekly_df['y_rolling'] = weekly_df['y'].rolling(4).mean()

plt.figure(figsize=(12, 5))
plt.plot(weekly_df['ds'], weekly_df['y'], label='Weekly Total_TP', alpha=0.5)
plt.plot(weekly_df['ds'], weekly_df['y_rolling'], label='4-week Rolling Avg', linewidth=2)
plt.title("Weekly Hydro Renewable Energy with Rolling Average")
plt.xlabel("Date")
plt.ylabel("Total_TP")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
corr = weekly_df.drop(columns=['ds']).corr()
sns.heatmap(corr[['y']].sort_values(by='y', ascending=False), annot=True, cmap='coolwarm')
plt.title("Correlation of Features with Hydro Energy (Total_TP)")
plt.tight_layout()
plt.show()
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error

#  1. Load Weekly Aggregated Data 
weekly_df = pd.read_csv("weekly_hydro_data.csv", parse_dates=["ds"])

#  2. Add Rolling Mean of Target 
weekly_df['y_rolling'] = weekly_df['y'].rolling(3).mean()

#  3. Drop NA caused by rolling 
weekly_df.dropna(inplace=True)

#  4. Train-Test Split 
train_df = weekly_df[weekly_df['ds'] < '2024-01-01']
test_df = weekly_df[weekly_df['ds'] >= '2024-01-01']

#  5. Initialize Prophet Model 
model = Prophet(
    seasonality_mode='multiplicative',
    yearly_seasonality=True,
    weekly_seasonality=True,
    changepoint_prior_scale=0.05
)

#  6. Add All Regressors from Heatmap (Attached the correlation values )
regressors = [
    'y_rolling',        # High positive correlation (0.93)
    'GWETPROF',         # Strong positive (0.78)
    'RH2M',             # Medium positive (0.51)
    'PRECTOTCORR',      # Low positive (0.22)
    'EVLAND',           # Slight negative (-0.05)
    'T2M'               # Strong negative (-0.51)
]

for reg in regressors:
    model.add_regressor(reg)

#  7. Fit the Model 
model.fit(train_df[['ds', 'y'] + regressors])

#  8. Forecast on Test Set 
future = test_df[['ds'] + regressors]
forecast = model.predict(future)

#  9. Predictions 
y_pred = forecast['yhat'].values
y_true = test_df['y'].values

#  10. Evaluation Metrics 
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

# FIXED MAPE Handling (remove division by zero)
y_true_np = np.array(y_true)
y_pred_np = np.array(y_pred)
non_zero_mask = y_true_np != 0
mape = np.mean(np.abs((y_true_np[non_zero_mask] - y_pred_np[non_zero_mask]) / y_true_np[non_zero_mask])) * 100

#  11. Print Results 
print(f"MAE: {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"MAPE: {mape:.2f}%")

#  12. Plot Actual vs Predicted 
plt.figure(figsize=(14, 6))
plt.plot(test_df['ds'], y_true, label='Actual', linewidth=2)
plt.plot(test_df['ds'], y_pred, label='Predicted (Prophet)', linestyle='--', color='red')
plt.title('Hydro Renewable Energy Forecast vs Actual (Prophet Model)\nAll Correlated Regressors from Heatmap')
plt.xlabel('Date')
plt.ylabel('Total_TP (Hydro Energy)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
