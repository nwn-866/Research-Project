```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error

# Load the dataset
file_path = "/Users/saluka/Downloads/CADairyProduction.csv"
data = pd.read_csv(file_path)

# Convert 'Year' and 'Month' columns to a datetime index
data['Date'] = pd.to_datetime(data['Year'].astype(str) + data['Month'], format='%Y%b')
data.set_index('Date', inplace=True)


milk_prod = data['Milk.Prod']

# Split the data into training and test sets
train = milk_prod[:-12]
test = milk_prod[-12:]


test.index = pd.date_range(start=train.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS')

# Fit Exponential Smoothing model
hw_model = ExponentialSmoothing(train, seasonal='add', seasonal_periods=12).fit()


hw_forecast = hw_model.forecast(steps=12)


plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Test Data', color='orange')
plt.plot(hw_forecast.index, hw_forecast, label='Predicted Data', color='green')
plt.title('Milk Production - Exponential Smoothing Forecast')
plt.xlabel('Date')
plt.ylabel('Milk Production')
plt.legend()
plt.show()


mse_hw = mean_squared_error(test, hw_forecast)
print(f'Mean Squared Error (Exponential Smoothing): {mse_hw}')
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit the SARIMA model
sarima_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)).fit(disp=False)

sarima_forecast = sarima_model.get_forecast(steps=12)
predicted_mean = sarima_forecast.predicted_mean
confidence_intervals = sarima_forecast.conf_int()

plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Test Data', color='orange')
plt.plot(predicted_mean.index, predicted_mean, label='Predicted Data (SARIMA)', color='blue')
#plt.fill_between(predicted_mean.index, confidence_intervals.iloc[:, 0], confidence_intervals.iloc[:, 1], color='blue', alpha=0.3)
plt.title('Milk Production - SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Milk Production')
plt.legend()
plt.show()

# Mean Squared Error for SARIMA
mse_sarima = mean_squared_error(test, predicted_mean)
print(f'Mean Squared Error (SARIMA): {mse_sarima}')
```

```python

plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Test Data', color='orange')
plt.plot(hw_forecast.index, hw_forecast, label='Predicted Data (Exponential Smoothing)', color='green')
plt.plot(predicted_mean.index, predicted_mean, label='Predicted Data (SARIMA)', color='blue')
#plt.fill_between(predicted_mean.index, confidence_intervals.iloc[:, 0], confidence_intervals.iloc[:, 1], color='blue', alpha=0.3)
plt.title('Milk Production - Model Comparison')
plt.xlabel('Date')
plt.ylabel('Milk Production')
plt.legend()
plt.show()

# Print the Mean Squared Errors for both models
print(f'Mean Squared Error (Exponential Smoothing): {mse_hw}')
print(f'Mean Squared Error (SARIMA): {mse_sarima}')
```

```python
import numpy as np

# Calculate differences
diff_hw = test.values - hw_forecast.values
diff_sarima = test.values - predicted_mean.values

# Calculate MAPE
mape_hw = np.mean(np.abs(diff_hw / test.values)) * 100
mape_sarima = np.mean(np.abs(diff_sarima / test.values)) * 100

results_df = pd.DataFrame({
    'Date': test.index,
    'Actual': test.values,
    'Forecast_Exponential_Smoothing': hw_forecast.values,
    'Difference_Exponential_Smoothing': diff_hw,
    'Forecast_SARIMA': predicted_mean.values,
    'Difference_SARIMA': diff_sarima
})

results_df.to_csv('forecast_comparison.csv', index=False)

print(f'MAPE (Exponential Smoothing): {mape_hw}%')
print(f'MAPE (SARIMA): {mape_sarima}%')

results_df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error


file_path = "/Users/saluka/Downloads/CADairyProduction.csv"
data = pd.read_csv(file_path)

# Convert 'Year' and 'Month' columns to a datetime index
data['Date'] = pd.to_datetime(data['Year'].astype(str) + data['Month'], format='%Y%b')
data.set_index('Date', inplace=True)

milk_prod = data['Milk.Prod']

# Split the data into training and test sets
train = milk_prod[:-12]
test = milk_prod[-12:]

# Ensure the test data is correctly aligned
test.index = pd.date_range(start=train.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS')

hw_model_full = ExponentialSmoothing(milk_prod, seasonal='add', seasonal_periods=12).fit()

sarima_model_full = SARIMAX(milk_prod, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)).fit(disp=False)

# Extend the forecasts till 2024
last_date = milk_prod.index[-1]
forecast_end_date = pd.to_datetime('2024-12-01')
extended_steps = (forecast_end_date.year - last_date.year) * 12 + (forecast_end_date.month - last_date.month)

# Make predictions
hw_forecast_extended = hw_model_full.forecast(steps=extended_steps)
sarima_forecast_extended = sarima_model_full.get_forecast(steps=extended_steps)
predicted_mean_extended = sarima_forecast_extended.predicted_mean
confidence_intervals_extended = sarima_forecast_extended.conf_int()

plt.figure(figsize=(12, 6))
plt.plot(milk_prod.index, milk_prod, label='Historical Data')
plt.plot(hw_forecast_extended.index, hw_forecast_extended, label='Forecast (Exponential Smoothing)', color='green')
plt.plot(predicted_mean_extended.index, predicted_mean_extended, label='Forecast (SARIMA)', color='blue')
plt.fill_between(predicted_mean_extended.index, confidence_intervals_extended.iloc[:, 0], confidence_intervals_extended.iloc[:, 1], color='blue', alpha=0.3)
plt.title('Milk Production Forecast - Extended Till 2024')
plt.xlabel('Date')
plt.ylabel('Milk Production')
plt.legend()
plt.show()

forecast_df = pd.DataFrame({
    'Date': hw_forecast_extended.index,
    'Forecast_Exponential_Smoothing': hw_forecast_extended.values,
    'Forecast_SARIMA': predicted_mean_extended.values,
    'Lower_CI_SARIMA': confidence_intervals_extended.iloc[:, 0].values,
    'Upper_CI_SARIMA': confidence_intervals_extended.iloc[:, 1].values
})



forecast_df.to_csv('extended_forecast_till_2024.csv', index=False)


```
