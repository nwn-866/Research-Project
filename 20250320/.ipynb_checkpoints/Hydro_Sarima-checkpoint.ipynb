```python
import pandas as pd

# Load Hydropower Generation Data
energy_data = pd.read_csv("energy_generation.csv", parse_dates=["Trading_Date"])
energy_data.set_index("Trading_Date", inplace=True)  # Set Date as Index

# Load Precipitation Data
precipitation_data = pd.read_csv("precipitation.csv", parse_dates=[["YEAR", "DOY"]])
precipitation_data.rename(columns={"YEAR_DOY": "Date"}, inplace=True)  # Rename merged column
precipitation_data.set_index("Date", inplace=True)  # Set Date as Index
```

```python
import pandas as pd

# Load Precipitation Data
precipitation_data = pd.read_csv("precipitation.csv")

# Convert YEAR and DOY (Day of Year) to a proper datetime format
precipitation_data["Date"] = pd.to_datetime(precipitation_data["YEAR"].astype(str) + precipitation_data["DOY"].astype(str), format="%Y%j")

# Set Date as the index
precipitation_data.set_index("Date", inplace=True)

# Drop YEAR and DOY as they are no longer needed
precipitation_data.drop(columns=["YEAR", "DOY"], inplace=True)

# Now Resample Precipitation Data to Monthly Sum
precipitation_monthly = precipitation_data.resample("M").sum()

# Ensure the date format matches energy data (set to month start)
precipitation_monthly.index = precipitation_monthly.index.to_period("M").to_timestamp()

# Print the final processed precipitation data
print(precipitation_monthly.head())
```

```python
# Load Energy Generation Data
energy_data = pd.read_csv("energy_generation.csv")

# Convert Trading_Date to datetime format
energy_data["Trading_Date"] = pd.to_datetime(energy_data["Trading_Date"], format="%Y-%m")

# Set as index
energy_data.set_index("Trading_Date", inplace=True)

# Print sample data
print(energy_data.head())
```

```python
# Merge datasets on the date index
merged_data = energy_data.join(precipitation_monthly, how="inner")

# Rename precipitation column for clarity
merged_data.rename(columns={"PRECTOTCORR": "Total_Precipitation"}, inplace=True)

# Print final dataset
print(merged_data.head())
```

```python
import matplotlib.pyplot as plt

# Plot Energy Generation Over Time
plt.figure(figsize=(12, 5))
plt.plot(merged_data.index, merged_data["Total_TP"], label="Energy Generation", color="blue")
plt.xlabel("Date")
plt.ylabel("Total Energy Generated")
plt.title("Hydropower Generation Over Time")
plt.legend()
plt.show()

# Plot Precipitation Over Time
plt.figure(figsize=(12, 5))
plt.plot(merged_data.index, merged_data["Total_Precipitation"], label="Precipitation", color="green")
plt.xlabel("Date")
plt.ylabel("Total Precipitation")
plt.title("Monthly Precipitation Over Time")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# ADF Test for Energy Generation
result = adfuller(merged_data["Total_TP"])
print(f"ADF Statistic (Energy): {result[0]}")
print(f"p-value (Energy): {result[1]}")

# ADF Test for Precipitation
result = adfuller(merged_data["Total_Precipitation"])
print(f"ADF Statistic (Precipitation): {result[0]}")
print(f"p-value (Precipitation): {result[1]}")
```

```python
merged_data["Energy_Diff"] = merged_data["Total_TP"].diff()

# Check stationarity again after differencing
from statsmodels.tsa.stattools import adfuller

result = adfuller(merged_data["Energy_Diff"].dropna())  # Drop NaN values after differencing
print(f"ADF Statistic (Energy after differencing): {result[0]}")
print(f"p-value (Energy after differencing): {result[1]}")
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot ACF (q)
plot_acf(merged_data["Energy_Diff"].dropna(), ax=axes[0])
axes[0].set_title("Autocorrelation (ACF)")

# Plot PACF (p)
plot_pacf(merged_data["Energy_Diff"].dropna(), ax=axes[1])
axes[1].set_title("Partial Autocorrelation (PACF)")

plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define the SARIMA model
sarima_model = SARIMAX(
    merged_data["Total_TP"],
    order=(1, 1, 1),  # (p, d, q)
    seasonal_order=(1, 1, 1, 12),  # (P, D, Q, s)
    enforce_stationarity=False,
    enforce_invertibility=False
)

# Fit the model
sarima_results = sarima_model.fit()

# Print summary
print(sarima_results.summary())
```

```python
# Predict next 12 months
forecast_steps = 12
forecast = sarima_results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=merged_data.index[-1], periods=forecast_steps, freq="M")

# Convert to DataFrame
forecast_df = pd.DataFrame({
    "Forecast": forecast.predicted_mean
}, index=forecast_index)

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(merged_data["Total_TP"], label="Observed Energy Generation")
plt.plot(forecast_df, label="Forecast", linestyle="dashed")
plt.legend()
plt.title("SARIMA Forecast for Energy Generation")
plt.show()
```
